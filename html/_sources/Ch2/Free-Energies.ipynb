# Free Energy 

The **free energy is the energy available to do useful work**. You probably learned about free energy in General Chemistry in the context of chemical reactions. In that context, exothermic reactions are reactions where the free energy change from the initial to final state is negative and endothermic processes are those where the free energy change is positive. 

In thermodynamics, we consider how the free energy is affected not by *chemical* processes but instead by *physical* ones.


## Free Energy Definition - The Combination of 1st and 2nd Laws:



It is convenient to define the free energies by beginning with the 1st and 2nd laws of thermodynamics. We have shown in previous lectures that the mass, energy, and entropy balance expressions can be written on a molar basis in the differential form as:

1. $\dfrac{dN}{dt}=\sum_k\dot{N}_k$


2. $\dfrac{dU}{dt}=\sum_k\big(\dot{N}\underline{H}\big)_k+\dot{Q}-P_{ext}\dfrac{dV}{dt}+\dot{W}$


3. $\dfrac{dS}{dt}=\sum_k\big(\dot{N}\underline{S}\big)_k+\dfrac{\dot{Q}}{T}+\dot{S}_{gen}$

If we combine (2) and (3), we can show that we get:

rearranging, we get

- $\dfrac{dU}{dt}=\sum_k\big(\dot{N}(\underline{H}-T\underline{S}\big)_k+T\dfrac{dS}{dt}-P_{ext}\dfrac{dV}{dt}-T\dot{S}_{gen}+\dot{W}$

Two useful cases emerge from this mess:

1. Open System Problems at Steady State

2. Closed Systems with Moving Boundaries

## Gibb's Free Energy

### Open Systems at Steady State

For an *open system* operating at *steady state* with *fixed boundaries*, we can write. 

$\quad 0=\sum_k\big(\dot{N}(\underline{H}-T\underline{S}\big)_k-T\dot{S}_{gen}+\dot{W}$

If we define the **Gibb's Free Energy** as $G=H-TS$, then the above can be simplified as:

$\quad 0=\sum_k\big(\dot{N}\underline{G}\big)_k-T\dot{S}_{gen}+\dot{W}$

#### Conceptual Meaning of Gibb's Free Energy

The form of the equation above explicitly links *work* and *entropy generation* to the change in *molar free energy* of a vapor/liquid as it undergoes some physical process. We can simplify things further for a single mass flow, $\dot{N}$. From (1), we know that $\dot{N}_2=-\dot{N}$ and therefore, we can write

$\quad 0=\dot{N}\big(\underline{G}_2-\underline{G}_1)-T\dot{S}_{gen}+\dot{W}$


This form of the equation is the most intuitive. If you remember anything from this class, remember these three maxims 

- Work done **by** the system $\dot{W}<0$ on the surroundings causes the a **decrease** in the **Gibb's Free Energy**, $\underline{G}_2<\underline{G}_1$, of the fluid. 


- Work done **on** the system $\dot{W}>0$ by the surroundings causes the an **increase** in the **Gibb's Free Energy**, $\underline{G}_2>\underline{G}_1$, of the fluid. 


- Entropy generation is always positive, $\dot{S}_{gen}>0$, therefore the interconversion of *work* to *free energy* and visa versa is always incomplete.
   
    - When work is done **on** on the system by the surroundings, this will result in a smaller increase in **Gibb's Free Energy** than for the *reversible* process.
    
    - When work is done **by** on the system by the surroundings, this will result in less work output for an equivalent change in **Gibb's Free Energy** than for the *reversible* process. 


We can imagine cyclically running a fluid through a compressor and turbine. The implication of the statements above is that through each cycle of compression and expansion, additional work must be supplied to make up for the accumulating losses associated with entropy generation. Therefore, perpetual motion machines that operate at steady state are impossible to construct and violate the **2nd Law of thermodynamics**. 

## Helmholtz Free Energy 

### Closed Systems with Moving Boundaries
For a closed system that has moving boundaries, we can group the time derivative terms on the L.H.S

- $\dfrac{d}{dt}\bigg(U-TS\bigg)=-P_{ext}\dfrac{dV}{dt}-T\dot{S}_{gen}$

If we define the **Helmholtz Free Energy** as $A=U-TS$, then the above can be simplified as:

- $\dfrac{dA}{dt}=-P_{ext}\dfrac{dV}{dt}-T\dot{S}_{gen}$'


#### Conceptual Meaning of the "Helmholtz Free Energy"

The form of the equation above explicitly links *work* and *entropy generation* to the change in extrinsic *free energy* of a vapor/liquid as it undergoes compression/expansion in a closed volume. Integrating both sides of this expression, we find that:

$\quad A_2-A_1=-P_{ext}\big(\Delta V) - \int_{1}^{2}T\dot{S}_{gen}dt$

where

$\quad \Delta V = V_{2}-V_{1}$


This form of the equation is the most intuitive. If you remember anything from this class, remember these three maxims 

- Work of expansion done **on** the surroundings $\Delta V>0$ by the system causes the a **decrease** in the **Helmholtz Free Energy**, $A_2<A_1$, of the fluid. 


- Work of compression done **by** the surroundings $\Delta V< 0$ on the system causes the an **increase** in the **Helmholtz Free Energy**, $A_2>A_1$, of the fluid. 


- Entropy generation is always positive, $\dot{S}_{gen}\ge 0$, therefore the interconversion of *boundary work* to *free energy* and visa versa is always incomplete.
   
    - When work is done **on** on the system by the surroundings, this will result in a smaller increase in **Helmholtz Free Energy** than for the **reversible** process.
    
    - When work is done **by** on the system by the surroundings, this will result in less work output for an equivalent change in **Helmholtz Free Energy** than for the **reversible** process. 


We can imagine a sealed piston cyclically compressing and expanding a fluid. The implication of the statements above is that through each cycle of compression and expansion, additional work must be supplied to make up for the accumulating losses associated with entropy generation. Therefore, perpetual motion machines that operate at steady state are impossible to construct and violate the **2nd Law of thermodynamics**.
