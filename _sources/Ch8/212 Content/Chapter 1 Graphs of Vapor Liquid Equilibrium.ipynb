## Different types of Separations

Engineers need to separate many types of mixtures to achieve multiple purposes.  One of the most common purposes is to remove a saleable product from a solvent or mixture of multiple products.  Another purpose could be to remove a pollutant or toxin from a waste stream.

In this course, we will focus on separating:
- liquids from vapors
- liquids from liquids

To achieve these types of separations, it is essential to exploit properties such as boiling point and solubility. 

It is important to understand these properties of mixtures at the operating conditions of a chemical process.  Mixtures exhibit different properties depending on which components are present and the temperature and pressure of the system.

We will begin our exploration of separations by considering separations of liquids and vapors.  To achieve these separations, it is important to understand how a given mixture behaves at equilibrium conditions.

## Foundational data and graphs of vapor-liquid equilibrium

### T-x-y diagrams

The Dortmund data bank also has T-x-y diagrams for the acetonitrile/1-butanol system

This figure is plotted at a constant temperature of 101.325 Pa.  The figure illustrates the composition of the liquid and vapor phases as the temperature is varied at this single pressure.  The lower line with points in the shape of circles is the composition of acetonitrile in the liquid phase (x).  The higher line with the points in the shape of triangles is the composition of acetonitrile in the vapor phase (y).

In the figure below, the region above the vapor phase composition line is labeled as vapor.  We know that this region is fully vapor because as T increases at a constant P and the system moves into a single phase, that phase will be vapor.  On the other hand, when the temperature decreases to regions below the liquid phase composition line, the single phase present is liquid.  Between the two lines, the mixture is in two phases.  

Note, when the temperature and pressure of a system dictate that it is in a single phase, it is *not possible* to separate the mixture using vapor-liquid separation techniques that we will soon discuss, like flash tanks and distillation columns.  Temperature and pressure must be chosen for any one mixture to enable vapor-liquid separation.

T-x-y diagrams tell us when to expect phase changes as a system moves across a range of temperatures at a constant pressure.  The T-x-y diagram can tell us which component is "lighter" or boils at a lower temperature at the pressure of the diagram.  Look at the T-x-y diagram for acetonitrile and 1-butanol above.  Which component would you guess has a lower boiling point?  

As a hint, consider what happens when the system *only* contains acetonitrile in the liquid phase such that x=1.  The system transitions to vapor at about 355 K.  On the other hand, when x=0 and the system only contains 1-butanol, the system would transition to vapor at 390 K.  

Therefore, acetonitrile is the more volatile component.  In fact, at 101.325 kPa (or 1 atm) acetonitrile's boiling point is 355 K; 1-butanol's is 390 K.  

What we really need T-x-y diagrams for is to tell us, for a system at a constant pressure and let's say a fixed liquid phase composition, what temperature do we need to be at to get to a two-phase system that is separable at all.  We can also use it to tell us what temperature we need to get to to achieve a target composition in the liquid and vapor phase.  This information is important for chemical engineers who need to produce, for example, products at a certain purity so that they can be sold or further processed.

Let's say we start with an acetonitrile, 1-butanol system at 360 K and a liquid phase composition of 0.3 mol acetonitrile/mol liquid phase.  The hollow orange circle on the chart below marks this point. When $x_{acetonitrile}$ is 0.3 at this temperature, the system is entirely liquid.  As the temperature heats up at constant composition, at T=366K, the system enters the two phase region.  With the tie line, we know at equilibrium at this temperature, when $x_{acetonitrile}$ is 0.3, $y_{acetonitrile}$ is 0.7.  What do we know about $x_{1-butanol}$ and $y_{1-butanol}$?  *Remember that in each phase, the sum of mole fractions must equal 1.*

The point where the x-phase composition line is reached at a constant liquid phase composition is called the **bubble point**.  Before that temperature is reached, the only phase is *liquid* but at that temperature, the very first bubble forms and the system becomes two-phase.  The x-composition line is in fact connecting all the bubble points of the system as the liquid phase mole fraction goes from 0 to 1.

What if we want a higher mole fraction of 1-butanol in the vapor phase upon separation?  1-butanol is the less volatile component. To push more of it into the vapor phase, we need to increase the temperature beyond 366 K.  Let's try 382 K.

At this temperature, we achieve a vapor phase mole fraction of acetonitrile of 0.3, which means that the vapor phase mole fraction of 1-butanol is 0.7. Do we need to start at a mole fraction of 0.3 at 366 K to achieve this composition?  No, an acetonitrile, 1-butanol system at 101.325 kPa and 382 K will always have this equilibrium concentration, regardless of the starting composition.

Another key concept to understand about T-x-y diagrams is the **dew point**.  At a constant mole fraction in the vapor region, as the temperature decreases at a constant pressure, eventually, the first drop of a liquid will appear and the system becomes two phases. Just as with the T-x-y diagram, it is possible to identify **bubble and dew points** on a P-x-y diagram.  The bubble point occurs when a system that is liquid phase only decreases in pressure such that a single bubble of gas forms.   The dew point occurs when a system is in a single, vapor phase and the pressure increases to the point where the first drop of liquid forms.  

In the figure below, the bubble point at an initial composition of 0.2 mole fraction acetonitrile in the liquid phase is shown at a pressure of 28 kPa, where the liquid phase composition line intersects the constant x=0.2 line.  We know from the tie line what the vapor phase composition is at the bubble point.  An analagous figure could be drawn for the dew point if we started at a vapor phase mole fraction of acetonitrile of 0.2 and 8 kPa.  The dew point would occur just at about 12 kPa.

### Key Take Aways

In this chapter we have reviewed three types of vapor-liquid equilibrium diagrams.  These diagrams help us understand the properties of systems we are trying to separate.  In particular, they help us understand the conditions at which the system contains two phases, which is necessary if we are to separate a binary system based on it separating into phases.  

- x-y diagrams tell us equilibrium compositions at either a set T or P.

- T-x-y and P-x-y diagrams tell us where we will have a two-phase region and what equilibrium compositions will be

- We can determine bubble and dew points from T-x-y and P-x-y diagrams.

- For a given system, there are interrelationships among the diagrams.
