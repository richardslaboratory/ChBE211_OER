## Forms of Kremser Equation:

**Form A:**
$$ \frac{y_{N+1} - y_j^*}{y_1 - y_1^*} = \left( \frac{1 - \left( \frac{L}{mV}^{N+1} \right)}{1 - \frac{L}{mV}} \right) $$

$$
N = \frac{\ln \left[ \left(1 - \frac{mV}{L} \right) \left(\frac{y_{N+1} - y_1^*}{y_1 - y_1^*} \right) + \frac{mV}{L} \right]}{\ln \left(\frac{L}{mV}\right)}
$$

eqn 12-21 Wankat 
  
Key points on McCabe-Thiele plot:
- $y_{N+1}$ is often given.
- $y_1$ could be calculated from operating line equation if $x_0$, $\frac{L}{V}$ known or could be specified
- $y_1^*$ could be specified where the $x = x_0$ intersects with the equilibrium line.
- If there were an eq. stag 0, it is the value of $y$ that would be in eqiulibrium with $x_0$.
- More importantly, this value of $y$ helps us calculate one of the $\Delta y $ we needed for Kremser equation derivation ( $\Delta y_0 $).



**Form B:**
  
eqn 12-24 Wankat

$$ \frac{y_{N+1} - y_{N+1}^*}{y_1 - y_1^*} = \left( \frac{L}{mV} \right)^N $$

$$ N = \ln \frac{\left[ \left( \frac{y_{N+1} - y_{N+1}^*}{y_1 - y_1^*} \right) \right] }{ \ln \left( \frac{L}{mV} \right) }$$

Form B still uses $y_1^*$, $ y_1$, and $y_{N+1}$

New: $y_{N+1}^*$ = mx_{N} + b $

Helps calculate last $ \Delta y $ of analytical approach.
