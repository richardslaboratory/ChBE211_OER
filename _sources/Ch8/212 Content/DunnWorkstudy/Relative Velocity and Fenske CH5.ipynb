Relative volatility:

$$ \alpha_{AB} = \frac{K_A}{K_B} = \frac{y_A/x_A}{y_B/x_B} $$

Relative volatility can be constant on a single stage:

$$ \frac{y_A}{y_B} = \alpha_{AB} \cdot \frac{x_A}{x_B} $$

Note notation on $\alpha$:

$$ \alpha_{AB} = \frac{K_A}{K_B} $$

$$ \alpha_{BA} = \frac{K_B}{K_A} $$

Could be:

$$ \alpha_{12} = \frac{K_1}{K_2} $$

$$ \alpha_{21} = \frac{K_2}{K_1} $$

etc.


# Fenske Equation

1. **Total Reflux → Look at reboiler**

## some figure

We would have overbars for L and V but are going to drop them here. The reboiler can be denoted as stage R or stage N+1.

Mass Bal around reboiler at total reflux (No bottom = product B = 0)

$$ V_{N+1} y_{N+1} = L_N x_N \text{(for a single component)}$$

Similar to a total condenser:

$$ V_{N+1} y_{N+1} = L_N x_N \quad \text{or} \quad y_{N+1} = x_N $$ for a component A, we can write this as:

$$ y_{A,R} = x_{A,N}, \quad y_{B,R} = x_{B,N} $$

At reboiler:

$$ \left[\frac{y_{A,R}}{y_{B,R}}\right] = \left[\frac{x_{A,N}}{x_{B,N}}\right] $$

→ Operating line is $y = x$  line. (recall our graphical method of finding $N_{min}$.

In general, for equilibrium:

$$ \left[\frac{y_A}{y_B}\right]_R = \alpha_R \left[\frac{x_A}{x_B} \right]_R $$

$$ \left[\frac{x_A}{x_B} \right]_N = \alpha_R \left[\frac{x_A}{x_B} \right]_R $$

Blue underlined terms are equilibrium substitution in the right-hand side of the above equation.

## some figure

Mass balance around stage N:

$$ V_{N+1} y_{N+1} + L_{N-1} x_{N-1} = L_N x_N + V_N y_N $$

$$ V_R y_R + L_{N-1} x_{N-1} = L_N x_N + V_N y_N $$


$$ \text{From before} \quad V_R y_R = L_N x_N $$

$$ L_{N-1} - x_{N-1} = V_N y_N $$

$$ \Rightarrow x_{N-1} = y_N $$

$$ \Rightarrow x_{A,N-1} = y_{A,N}, \quad x_{B,N-1} = y_{B,N} $$


## Equilibrium relationship at stage N

$$ \left[\frac{y_A}{y_B}\right]_N = \alpha_N \left[\frac{x_A}{x_B} \right]_N $$

$$ \left[\frac{x_A}{x_B}\right]_{N-1} = \alpha_N \left[ \frac{x_A}{x_B} \right]_N $$

$$ \left[\frac{x_A}{x_B} \right]_N = \alpha_R \left[ \frac{x_A}{x_B} \right]_R $$

$$ \left[\frac{x_A}{x_B} \right]_{N-1} = \alpha_N \alpha_R \left[\frac{x_A}{x_B} \right]_R $$

Can keep walking up the column alternating between solving mass bal, using eq. relationship to get to the point

$$ \left[ \frac{x_A}{x_B} \right]_D = \alpha_1, \alpha_2, \ldots, \alpha_{N-1},\alpha_N, \alpha_R \left[\frac{x_A}{x_B} \right]_R $$

At the distillate stream

The definition of a geometric mean is:

$$ \alpha_{AB} = \left[\alpha_1 \alpha_2 \alpha_3 \ldots \alpha_{N-1} \alpha_{N} \alpha_R \right]^{1/N} \rightarrow ??? N = N_{\text{min}} $$

$$ \alpha_{AB} N_{\text{min}} = \left[\alpha_1 \ldots \alpha_R \right] $$

We will use this to characterize the geometric average relative volatility

$$ \left[\frac{x_A}{x_B} \right]_D = \alpha_{AB}N_{\text{min}} \left[\frac{x_A}{x_B} \right]_R $$

$$ N_{\text{min}} = \frac{\ln \left[ \left[ \frac{x_A}{x_B}\right]_D / \left[ \frac{x_A}{x_B}\right]_R\right]}{\ln \alpha_{AB}} $$

One form of the Fenske Equation works best for constant relative volatility.
  
To use in terms of fractional recovery...

$$ N_{\text{min}} = \frac{ \ln \left[ \frac{f_{?A} \cdot \cancel{F_{?A}}}{f_{?B} \cdot \cancel{F_{?B}}}  \cdot \frac{f_{B_1B} \cdot \cancel{F_{2B}}}{f_{B_1A} \cdot \cancel{F_{2A}}} \right] } {\ln \alpha_{AB}}$$  
  
$$ = \frac{\ln \left[ \frac{f_{D_1A} \cdot f_{B_1B}}{(1 - f_{B_1B})(1 - f_{D_1A})} \right]}{\ln \alpha_{AB}} $$
