## Thermodynamics of Mixtures

### Characteristics of Equilibrium

At equilibrium, the following are equal for each component
- Temperature
- Pressure
- Fugacity

We encountered pure component fugacity and fugacity coefficient in 211. As a reminder:

Fugacity:
$$f = P\,exp\left \{ \frac{\underline{G}(T,P)-\underline{G}^{IG}(T,P)}{RT} \right \} \\ = P\,exp\left \{ \frac{1}{RT}\int_{0}^{P}\left ( \underline{V} - \frac{RT}{P}dP \right ) \right \}$$

Fugacity Coefficient:
$$\Phi = \frac{f}{P}= exp\left \{ \frac{\underline{G}(T,P)-\underline{G}^{IG}(T,P)}{RT} \right \} \\= exp\left \{ \frac{1}{RT}\int_{0}^{P}\left ( \underline{V} - \frac{RT}{P}dP \right ) \right \}$$

Yet, in these sytems, x = 1 and y = 1. The question was just how much of the pure component was in each phase.

### Defining Fugacity

Oxford Dictionary: Fugacity is a thermodynamic property of a *real gas* which if **substituted** for the pressure or partial pressure in the equations for an ideal gas gives equations applicable to the real gas

From 211: It is not immediately obvious how equilibrium conditions are determine from a volumetric equation of state. We needed something else to make this connection. Hence, the concept of fugacity was invented. Fugacity is a mathematical relationship representing how fugitive (volatile) a fluid is.

### Methods to Determine Fugacities in Mixtures

There are three methods we can use to determine fugacities in mixtures:
- Equations of State 
- Data 
- Predictive Models 

#### Equations of State (EOS)

Equations of state relate **state variables: P, T, V**. They have undergone many changes and advancements to better characterize chemical behavior. In the case of separations, EOSs need to help us understand the behavior of each component in a mixture. To do this well, they need to capture things like interactions among molecules at different conditions.

Note that developing EOS is a long-standing area of research.  As Kontogeorgis et al. (2020) summarize, this research has spanned centuries (https://www.sciencedirect.com/science/article/pii/S259014002030006X).

### Mixtures May or May Not Be Ideal

Mixtures may or may not behave ideally. However, we can typically assume a gaseous mixture to be ideal when the system is:
- At low temperature (T), and low pressure (P)
- The mixture is very dilute; interactions among molecules do not contribute significantly to total system energy

When we have an ideal mixture system, the ideal gas law applies $$PV = nRT$$

Additionally, there is no change in properties upon mixing (like internal energy, etc).

**Key point:** in an ideal mixture, the partial pressure of a species is equal to its mole fraction times the total pressure $P_i = x_iP$)

Unfortunately, most mixtures are not ideal. But, we still need a way to estimate their thermodynamic properties.

### Partial Molar Properties

A property that is a function of (T, P, and $\underline{x}$) is a function of temperature, pressure, and all of the mole fractions. That is, the function depends on the relative amounts of every component in a phase, not just one.

The **overbar** on a property means it is a partial molar property.

Example:
$$V = x_1\overline{V_1} + x_2\overline{V_2}$$

### Excess Mixing Properties Account for Differences Between Ideal and Real Mixtures

#### Example: Volume at Constant T and P

##### Ideal:

- Partial molar volume of i in an ideal mixture equals its molar property $$\overline{V}^{IM} = \underline{V_i}$$
- Change in volume upon mixing of multiple components is 0
$$ \Delta_{mix}\;\underline{V}^{IM} = 0$$
- Molar volume of an ideal mixture is the sum of the molar volumes of each of its components times their molar volumes
$$ \underline{V}^{IM} = \sum_{i=1}^{c} x_i\underline{V_i}$$

##### Real:

- Excess partial molar volume of i in a real mixture is its partial molar volume at this T and P less its molar volume under ideal conditions
$$ \overline{V}_i^{ex} = \overline{V}_i - \underline{V}_i$$

- Excess molar volume of a real mixture is the change upon mixing of its molar volume

$$ \underline{V}^{ex} = \Delta_{mix}\; \underline{V} $$

- Combining, we get:

$$ \underline{V} = \sum_{i=1}^{c} x_i\underline{V_i} + \Delta_{mix}\; \underline{V} = \underline{V}^{IM} + \underline{V}^{ex}$$

### At equilibrium: Gibbs free energies of each phase are equal

This equilibrium criterion gives us a way to work towards understanding the composition of each phase.

Using the relationships between Gibbs free energy and fugacity (see Section 9.2), it is possible to demonstrate that fugacities of components in different phases are also equal at equilibrium.

$$ \overline{f}_i^I = \overline{f}_i(T, P, \underline{x}^I) = \underline{f}_i(T, P, \underline{x}^{II}) = \overline{f}_i^{II} $$

<span style="color:#DA7407">**At equilibrium, the fugacities of each component must be equal in the two phases.**</span>

This is what makes it so important to be able to find fugacities and, eventually, relate them to composition. We want to be able to predict composition at different T and P, so we know if we are able to hit our desired concentrations, for example, of products.

## Fugacity of a Single Component in Multicomponent Systems

- Fugacity of component *i* in a mixture:

$$ \overline{f}_i(T, P, \underline{x}) = P\,exp\left \{ \frac{\overline{G}_i(T,P,\underline{x})-\underline{G}_i^{IG}(T,P,\underline{x})}{RT} \right \} $$

- Fugacity coefficient for a component *i* in a mixture:

$$ \Phi_i = \frac{\overline{f}_i}{x_iP} $$

Fugacity is introduced to handle non-ideality.

<span style="color:#DA7407">**As pressures lower towards 0, the system will act increasingly ideal.**</span>

$$ \overline{f}_i \rightarrow x_iP \equiv P_i $$


## When compressability is approximately linear as a function of mole fraction use, Lewis-Randall Rule for vapor phase fugacity

Compressibility:

$$ Z = \frac{P}{NRT} $$

For ideal gases, $Z = 1$. Z could be a linear function of a mole fraction over a P range for some gaseous mixtures. In this case, we can use the Lewis-Randall Rule to estimate a component's vapor phase fugacity.

<span style="color:#DA7407">**The Lewis-Randall Rule**</span>
$$ \overline{f}_i^V(T,P,\underline{y}) = y_if_i^V(T,P) $$

We need to know pure component vapor phase fugacity and vapor phase mole fraction.

## Alternative: Use Truncated Virial EOS for dilute gases

One option is the *"truncated virial"* EOS.

Virial EOS describes the behavior of dilute gases and are typically written as an expansion of compressibility, $Z$.

Low pressure, truncated virial equation of state:

$$ \frac{P\underline{V}}{RT} = 1 + \frac{B_{mix}(T,\underline{y})}{\underline{V}} = Z_{mix} $$

where

$$ B_{mix}(T,\underline{y}) = \sum_i \sum_j{y_iy_jB_{ij}(T)}$$

B is a function of T, and

$$B_{ij} = B_{ji}$$

<span style="color:#DA7407">**Equation for calculating fugacity of a component in a gaseous mixture using Virial EOS:**</span>

$$ ln\left (\frac{\overline{f}_i^V}{y_iP} \right) = \frac{2P}{Z_{mix}RT} \sum_j{y_jB_{ij}(T)-ln(Z_{mix})}$$

where

$$ Z_{mix} = \frac{1}{2} \left ( 1 + \sqrt{1 + \frac{4B_{mix}}{RT}} \right ) $$

## Summary: single component fugacity, vapor phase

We now have two ways to estimate a single component's fugacity in a mixture of gases:
- Lewis-Randall rule: relies on one type of behavior (linear changes in compressibility as a function of mole fraction)
- Virial Equation of State: best for lower pressures

## Liquids

We want to be able to characterize liquid mixtures and first need to think about how they behave when they are ideal and not ideal.

$$ \overline{H}_i^{IM}(T,P,\underline{x}) = \underline{H}_i(T,P)$$

$$ \overline{V}_i^{IM}(T,P,\underline{x}) = \underline{V}_i(T,P)$$

*Interpreting these equations:* In an ideal mixture, for any compoennt, its partial molar property (enthaly, volume, etc) will be equal to its property per mole. This is true at any T, P composition.

$$ \Delta_{mix}V^{IM}(T,P,\underline{x}) = 0 $$
$$ \Delta_{mix}H^{IM}(T,P,\underline{x}) = 0 $$

*Interpreting these equations:* There is no $V$ or $H$ change when an ideal mixture forms from pure components.

## Key Concept: Activity Coefficient

- When an EOS is available for **both** the pure and mixed fluid, we have enough information to characterize the system
    - Peng-Robinson could work
- Yet, this is rare, so we bring in another relationship developed through thermo theory (the activity coefficient, $\gamma_i$)

<p style="text-align: center;"> <span style="color:#7B2CBF">$\overline{f}_i^L(T,P,\underline{x})$</span>
$ = x_i$ <span style="color:#F41101">$\gamma_i(T,P,\underline{x})$</span> <span style="color:#78A57E">$f_i^L(T,P)$</span></p>

- <span style="color:#7B2CBF">Fugacity of component *i* in a mixture</span>
- <span style="color:#F41101">Activity coefficient</span>
- <span style="color:#78A57E">Pure liquid fugacity</span>

If the mixtures were **ideal**, the fugacity of the component would be equal to its mole fraction times the pure component fugacity (<span style="color:#F41101">activity coefficient = 1</span>)

## Fugacity Coefficients Also Relate to Excess Gibbs Free Energy

Recall:

$$ \overline{G}_i^{ex} = RTln(\gamma_i(T,P,\underline{x}) $$

It is possible to measure $\overline{G}_i^{ex}$ experimentally and then estimate the activity coefficient. Based on these relationships (see Section 9.3), it is possible to derive the <span style="color:#DA7407">**Gibbs-Duhem Equation**</span> for activity coefficients:

For a single T, P:

$$ x_1 \left (\frac{\partial ln\gamma_1}{\partial x_1} \right)_{T,P} + x_2 \left (\frac{\partial ln\gamma_2}{\partial x_2} \right)_{T,P} = 0$$ 

Over a *narrow* temperature range, we can adjust for changes in T:

$$\gamma_i(T_2,P,\underline{x}) = \gamma_i(T_1,P,\underline{x})exp \left [ \frac{\overline{H}_i^{ex}}{R} \left (\frac{1}{T_2}-\frac{1}{T_1} \right ) \right ]$$


## Liquid Mixtures

Approaches:

- Equations of state (example: Peng-Robinson)
- When EOS fails, we need to turn to other methods

<p style="text-align: center;"> <span style="color:#7B2CBF">$\overline{f}_i^L(T,P,\underline{x})$</span>

We are considering simple mixtures - mixtures in which all components are in the liquid phase at T and P of mixing. We would take a different approach to characterizing mixtures with dissolved solids or gases.

The change in Gibbs free energy when a simple mixture is formed is:

##### Simple system

The simplest polynomial that represents $\underline{G}^{ex}$ is $Ax_1x_2$

From this relationship, we can derive the one-constant Margules activity coefficients.

$$ RT\,ln\gamma_1 = Ax_2^2 $$
$$ RT\,ln\gamma_2 = Ax_1^2 $$

What happens to $\gamma_1$ as $x_1$ moves towards 1? (Assume $A/RT = 1$)

$\gamma_1$ moves towards 1, which means the fugacity of component in the mixture is about equal to the pure liquid fugacity.

<p style="text-align: center;"> <span style="color:#7B2CBF">$\overline{f}_i^L(T,P,\underline{x})$</span>
$ = x_i$ <span style="color:#F41101">$\gamma_i(T,P,\underline{x})$</span> <span style="color:#78A57E">$f_i^L(T,P)$</span></p>

As species *i* becomes more dilute, its behavior increasingly deviates from the pure component's.

##### More complex systems

For more complex systems, the simplest polynomial does not do a good job characterizing the system.

We need to use a polynomial with more terms to capture behavior. Use Redlich-Kister expansion:

$$\underline{G}^{ex} = x_1x_2\{A + B(x_1 - x_2) + C(x_1 - x_2)^2 + \cdots \}$$

<span style="color:#DA7407">**Two Constant Margules Equations:**</span>

$$ RT\,ln\gamma_1 = \alpha_1x_2^2 + \beta_1x_2^3 $$ 
$$ RT\,ln\gamma_2 = \alpha_1x_2^2 + \beta_1x_1^3 $$

where

$$ \alpha_i = A + 3(-1)^{i+1}B \\ \beta_i = 4(-1)^iB $$

Redlich-Kister expansion assumes completely random mixtures of molecules form.

Other expansion methods yield the <span style="color:#DA7407">**van Laar Equations:**</span>

$$ln(\gamma_1) = \frac{\alpha}{\left [ 1 + \frac{\alpha}{\beta}\frac{x_1}{x_2} \right ]^2}; \;\; \alpha = 2q_1a_{12}$$

$$ln(\gamma_2) = \frac{\beta}{\left [ 1 + \frac{\beta}{\alpha}\frac{x_2}{x_1} \right ]^2}; \;\; \beta = 2q_2a_{12} $$

$q$ is a measure of the volume of a molecule (could be "b" from van der Waals equation)

Usually parameters are found by fitting equations to experimental activity coefficient data.

<span style="color:#DA7407">**van Laar equations when data are limited:**</span>

$$ \alpha = \left (1 + \frac{x_2ln(\gamma_2)}{x_1ln(\gamma_1)} \right)^2 ln(\gamma_1) $$

$$ \beta = \left (1 + \frac{x_1ln(\gamma_1)}{x_2ln(\gamma_2)} \right)^2 ln(\gamma_2) $$

Advantage: if we have $\gamma_1$ and $\gamma_2$ at a single mole fraction, we can evaluate $\alpha$ and $\beta$ and then calculate activity coefficients at different mole fractions.

##### Why do we need so many models?

Different approaches to modeling treat interactions among molecules differently.

Redlich-Kister expansion assumes:
- Completely random mixtures are formed. The environment around any one molecule reflects the ratio of mole fractions in the entire mixture.
- In reality, any one molecule's "neighborhood" may not be so simple to characterize
- Other models account for differences in how molecules interact based on size and energy
    - Local composition is not equal to bulk composition in these cases
    
    
<span style="color:#DA7407">**Two-parameter Wilson Equation:**</span>

The two-parameter Wilson equation accounts for a local environment around molecules that may not reflect the bulk composition.

$$ \frac{\underline{G}^E}{RT} = -x_1ln(x_1 + x_2\Lambda_{12}) - x_2ln(x_2 + x_1\Lambda_{21}) $$

$$ln(\gamma_1) = -ln(x_1 + x_2\Lambda_{12}) + x_2 \left [ \frac{\Lambda_{12}}{x_1 + x_2\Lambda_{12}} - \frac{\Lambda_{21}}{x_1\Lambda_{21} + x_2} \right ] $$

$$ln(\gamma_2) = -ln(x_2 + x_1\Lambda_{21}) + x_1 \left [ \frac{\Lambda_{12}}{x_1 + x_2\Lambda_{12}} - \frac{\Lambda_{21}}{x_1\Lambda_{21} + x_2} \right ] $$

At infinite dilution, species 1 or 2 become infinitely dilute such that $x_1$ or $x_2$ is equal to zero

$$ ln(\gamma_1^\infty) = -ln(\Lambda_{12}) + 1 - \Lambda_{21}; \;\; ln(\gamma_2^\infty) = -ln(\Lambda_{21}) + 1 - \Lambda_{12} $$

<span style="color:#DA7407">Parameters $\Lambda_{ij}$ account for the differences in size and interaction energies of mixture components. These essentially weight the mole fractions.</span>


<span style="color:#DA7407">**Non-Random Two-Liquid (NRTL) Model:**</span>

Another model that lets us account for different local environments around molecules is the non-random two-liquid (NRTL) model:

$$ \frac{\underline{G}^E}{RT} = x_1x_2 \left (\frac{\tau{21}G_{21}}{x_1 + x_2G_{21}} + \frac {\tau_{12}G_{12}}{x_2 + x_1G_{12}} \right ) $$

$$ ln(G_{12}) = -\alpha\tau_{12}; \;\; ln(G_{21}) = -\alpha\tau_{21} $$

<span style="color:#DA7407">$\tau$ replaces $\Lambda$ as the parameter reflecting size/energies of molecules </span>

$$ ln(\gamma_1) = x_2^2 \left [ \tau_{21} \left ( \frac{G_{21}}{x_1 + x_2G_{21}} \right )^2 + \frac{\tau_{12}G_{12}}{(x_2 + x_1G_{12})^2} \right ] $$

$$ ln(\gamma_2) = x_1^2 \left [ \tau_{12} \left ( \frac{G_{12}}{x_2 + x_1G_{12}} \right )^2 + \frac{\tau_{21}G_{21}}{(x_1 + x_2G_{21})^2} \right ] $$

<span style="color:#DA7407">**Universal Quasichemical Model (UNIQUAC):**</span>

Based on statistical mechanics:

$$ \frac{\underline{G}^{ex}}{RT} = \frac{\underline{G}^{ex}(combinatorial)}{RT} + \frac{\underline{G}^{ex}(residual)}{RT} $$

Combinational accounts for differences in molecular shapes and sizes.
Residual accounts for differences in molecular energies.

$$ ln(\gamma_i) = ln(\gamma_i(combinatorial)) + ln(\gamma_i(residual)) $$
$$ ln(\gamma_i(combinatorial)) = ln\frac{\phi_i}{x_i} - \frac{z}{2}q_iln\frac{\phi_i}{\theta_i} + l_i - \frac{\phi_i}{x_i}\sum_{j}{x_jl_j}$$

$$ ln(\gamma_i(residual)) = q_i \left [ 1 - ln \left (\sum_{j}{\theta_j \tau_{ji}} \right ) - \sum_{j}{\frac{\theta_j\tau_{ij}}{\sum_{k}\theta_k\tau_{kj}}} \right ] $$

$$l_i = \frac{(r_i - q_i)z}{2-(r_i - 1)}$$

where $r_i$ = volume parameter for species *i*, $q_i$ = surface area parameter for species *i*. *Derived from molecular structure information.* $\tau$ are the adjustable parameters.

Area fraction of species *i*: $$\theta_i = \frac{x_iq_i}{\sum_j{x_j}{q_j}} $$

Volume fraction of species *i*: $$\phi_i = \frac{x_ir_i}{\sum_{j}{x_jr_j}} $$

What if we have no data, which is entirely possible given the wide range of mixtures we might have?

We can use predictive methods based on theory.

We will talk about one of these <span style="color:#DA7407">**UNIFAC - UNIquac Functional-group Activity Coefficient**</span>

This method is based on UNIQUAC:

$$ ln(\gamma_i) = ln(\gamma_i(combinatorial)) + ln(\gamma_i(residual)) $$
$$ ln(\gamma_i(combinatorial)) = ln\frac{\phi_i}{x_i} - \frac{z}{2}q_iln\frac{\phi_i}{\theta_i} + l_i - \frac{\phi_i}{x_i}\sum_{j}{x_jl_j}$$

This term relies on having data to fit.

$$ln(\gamma_i(combinatorial)) = ln\frac{\phi_i^\prime}{x_i} + 1 - \frac{\phi_i^\prime}{x_1} - \frac{z}{2}q_i \left ( 1 + ln\frac{\phi_i}{\theta_i} - \frac{\phi_i}{\theta_i} \right ) $$

$$\phi_i^\prime = \frac{x_ir_i^{3/4}}{\sum_{j}{x_jr_j^{3/4}}} $$

$$ ln(\Gamma_k) = Q_k \left [ 1 - ln \left (\sum_m{\theta_m \Psi_{mk}} \right ) - \sum_m{\frac{\theta_m\Psi_{mk}}{\sum_n{\theta_n\Psi_{nm}}}} \right ] $$

Surface area fraction of group m:

$$ \theta_m = \frac{X_mQ_m}{\sum_n{X_nQ_n}} $$

where $X_m$ is the mole fraction of group M in mixture

$$\Psi_m = exp \left [ \frac{-(u_{mn}-u_{nm})}{kT} \right ] $$

$u_{mn}$ measures the interaction energy between groups *m* and *n*. These are unknowns that need to be determined through a lot of experimental data. Assumed to be the same regardless of the molecule.

$$ln(\gamma_i(residual)) = \sum_k{v_k^{(i)}\left [ln(\Gamma_k) - ln(\Gamma_k^{(i)})\right ]}$$

where $v_k^{(i)}$ is the number of *k* groups in species *i*. If we have a pure species *i*, the activity coefficient will be zero.

## Key Points about UNIQUAC and UNIFAC

- One is predictive (UNIFAC) and the other is based on correlations with data (UNIQUAC)
- Use UNIFAC when there are no data for the mixture we are working with
- Both methods account for the size/shape and interaction energy of molecules within mixtures
