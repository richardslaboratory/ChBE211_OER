# Computing Thermal Variables from an Equation of State

Last time, we derive open and closed system expressions that for *isotherml* paths allow us to compute changes in *thermal variables* from volumetric variables alone. The sets of equations that we derived are shown below:

$\quad d\underline{S}=\dfrac{C_V}{T}dT+
\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}d\underline{V}$

$\quad d\underline{U}=C_{V}dT+
\bigg(T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-P\bigg)d\underline{V}$


$\quad d\underline{S}=\dfrac{C_P}{T}dT-
\bigg(\dfrac{\partial \underline{V}}{\partial T}\bigg)_PdP$

$\quad d\underline{H}=C_{P}dT+
\bigg(\underline{V}-T\bigg(\dfrac{\partial \underline{V}}{\partial T}\bigg)_{P}\bigg)dP$

We saw in Ch 3.that we rarely encounter *isothermal* pathways in practice and so we need to develop strategies that allow for such computations for any combination of volumetric property change. Of course, we can see from the above equations that isothermal pathways were convenient because we did not need heat capacity data for the substance, but for other property changes, we will need this information. Unfortunately, just like thermal equations of state, we do not have access to heat capacity data at all state points. Why then were these derivatives useful?

## Calculation Pathways

The criteria for the path that we chose should 

1. Minimize the amount of heat capacity data that needs to be tabulated 

2. Allow for direct application of volumetric equations of state 

3. Not require any other property data

The path that satisfies these three criteria is to go through the *ideal gas state* which we will illustrate below:

### Pathways for Closed System Problems:

If $\underline{U}=\mathit{f}\big(\underline{V},T\big)$, and we need to compute property changes from $T_1, \underline{V}_1$ to $T_2, \underline{V}_2$, then we will perform a 3 step calculation:

1. Isothermal expansion from $\underline{V}_{1}$ to $\underline{V}=\infty$ at $T_1$

2. Temperature change from $T_1$ to $T_2$

3. Isothermal compression from $\underline{V}=\infty$ to $\underline{V}_2$ at $T_2$

This 3-step path takes us through the ideal gas state, recall that $\underline{V}\rightarrow\infty$ corresponds to the infinitely expanded limit where all substances behave as ideal gases. In this way, step (2) can be performed in the ideal gas state where the *thermal variables* are only a function of temperature and the *ideal gas heat capactiies* are easily tabulated for many substances. 

The total change in internal energy $\underline{U}_{1\rightarrow2}$ and $\underline{S}_{1\rightarrow2}$ for this three step pathway can be expressed as:

$\quad \Delta\underline{S}_{1\rightarrow2}=\int_{T_1}^{T_2}\dfrac{C_V^{IG}}{T}dT+
\int_{\underline{V}_1}^{\infty}\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}d\underline{V}
+\int_{\infty}^{\underline{V}_2}\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}d\underline{V}$


$\quad \Delta\underline{U}_{1\rightarrow2}=\int_{T_1}^{T_2}C_V^{IG}dT+
\int_{\underline{V}_1}^{\infty}\bigg[T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-P\bigg]_{T_1}d\underline{V}
+\int_{\infty}^{\underline{V}_2}\bigg[T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-P\bigg]_{T_2}d\underline{V}$





### Pathways for Open System Problems:

If $\underline{H}=\mathit{f}\big(P,T\big)$, and we need to compute property changes from $T_1, P_1$ to $T_2, P_2$, then we will perform a 3 step calculation:

1. Isothermal expansion from $P_{1}$ to $P\rightarrow0$ at $T_1$

2. Temperature change from $T_1$ to $T_2$

3. Isothermal compression from $P\rightarrow 0$ to $P_2$ at $T_2$

Just like the close system path, this path goes through the ideal gas state, where $P=0$. Again we can use ideal gas heat capacities for the purpose of computation and develop expressions for $\underline{H}_{1\rightarrow2}$ and $\underline{S}_{1\rightarrow2}$

$\quad \Delta\underline{S}_{1\rightarrow2}=\int_{T_1}^{T_2}\dfrac{C_P^{IG}}{T}dT
-\int_{P_1}^{0}\bigg(\dfrac{\partial \underline{V}}{\partial T}\bigg)_{P}dP
-\int_{0}^{P_2}\bigg(\dfrac{\partial \underline{V}}{\partial T}\bigg)_{P}dP$

$\quad \Delta\underline{H}_{1\rightarrow2}=\int_{T_1}^{T_2}C_P^{IG}dT+
\int_{P_1}^{0}\bigg[\underline{V}-T\bigg(\dfrac{\partial \underline{V}}{\partial T}\bigg)_{P}\bigg]_{T_1}dP
+\int_{0}^{P_2}\bigg[\underline{V}-T\bigg(\dfrac{\partial \underline{V}}{\partial T}\bigg)_{P}\bigg]_{T_2}dP$


## Departure Functions

Armed with the equations above, a volumetric equation of state, and an ideal gas heat capacity for a substance of interest, it is possible calculate all thermal properties of interest and even construct property diagrams for these substances. Before we demonstrate the construction of property diagrams using these expresisons, we will take a small *departure* to discuss **departure functions**. 

A **departure function** is a function that describes the deviation of real gas behavior from ideal gas behavior. We will see that departure functions are used to compute fugacities for vapor-liquid equilibrium calculations and for many other purposes.

Departure functions are frequently denoted with a superscript **R**. To compute a departure function, we must first calculate the property change for the path taken by an ideal gas. We will do this for the paths described above.

### Departure Functions for Closed System Paths:

$\quad \Delta\underline{S}_{1\rightarrow2}^{IG}
=\int_{T_1}^{T_2}\dfrac{C_V^{IG}}{T}dT
+Rln\bigg(\dfrac{\underline{V}_2}{\underline{V}_1}\bigg)$

$\quad \Delta\underline{U}_{1\rightarrow2}^{IG}=\int_{T_1}^{T_2}C_V^{IG}dT$

This makes the **departure functions**

$\quad \Delta\underline{S}_{1\rightarrow2}^{R}
=\Delta\underline{S}_{1\rightarrow2}
-\Delta\underline{S}_{1\rightarrow2}^{IG}
=
\int_{\underline{V}_1}^{\infty}\bigg[\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}\bigg]_{T_1}d\underline{V}+
\int_{\infty}^{\underline{V}_2}\bigg[\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}\bigg]_{T_2}d\underline{V}
$





$\quad \Delta\underline{U}_{1\rightarrow2}^{R}
=\Delta\underline{U}_{1\rightarrow2}
-\Delta\underline{U}_{1\rightarrow2}^{IG}
=\int_{\underline{V}_1}^{\infty}\bigg[T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-P\bigg]_{T_1}d\underline{V}+
\int_{\infty}^{\underline{V}_2}\bigg[T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-P\bigg]_{T_2}d\underline{V}$


### Departure Functions for Open System Paths:

$\quad \Delta\underline{S}_{1\rightarrow2}^{IG}=\int_{T_1}^{T_2}\dfrac{C_P^{IG}}{T}dT-Rln\bigg(\dfrac{\underline{P}_2}{\underline{P}_1}\bigg)$

$\quad \Delta\underline{H}_{1\rightarrow2}^{IG}=\int_{T_1}^{T_2}C_P^{IG}dT$

### Useful form of Departure Function:

The derivative with respect to $d\underline{V}$ is not so convenient for the volumetric equations of state that we routinely have available. Therefore, it is often preferred to rearrange the expression above for the departure function of the enthalpy to a more *useful form*. We proceed by recognizing that:

$\quad dP=\dfrac{1}{\underline{V}}d(P\underline{V})-\dfrac{P}{\underline{V}}d\underline{V}$

using the above, it can be shown that

$\quad \Delta\underline{S}_{1\rightarrow2}^{R}
=R\ln{\dfrac{Z_2}{Z_1}}+ 
+\int_{\underline{V}_1}^{\infty}\bigg[\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-\dfrac{R}{\underline{V}}\bigg]_{T_1}d\underline{V}
+
\int_{\infty}^{\underline{V}_2}\bigg[\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-\dfrac{R}{\underline{V}}\bigg]_{T_2}d\underline{V}$


$\quad \Delta\underline{H}_{1\rightarrow2}^{R}
=RT_2(Z_2-1)-RT_1(Z_1-1)+
\int_{\underline{V}_1}^{\infty}\bigg[T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-P\bigg]_{T_1}d\underline{V}+
\int_{\infty}^{\underline{V}_2}\bigg[T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-P\bigg]_{T_2}d\underline{V}$



### Illustration Problem:

Steam undergoes a Joule-Thomson Expansion from $P=25$ $bar$ and $T=300^{o}C$ to $P=1$ $bar$ and $T=?$. Find the temperature at the outlet for:

1. An ideal gas

2. A van der Waals gas. 

    - $a=0.5542 Pa\cdot m^6/mol^2$
    
    - $b=3.051\times 10^{-5} m^3/mol $
    
    - $C_p^{IG} [J/(mol\cdot K)]=32.218+0.01449 T$
    
3. Steam.

Compare the results. Is steam well approximated by an ideal gas?

#### First let's plot an isotherm at this temperature:

```python
import matplotlib.pyplot as plt
import numpy as np

R=8.314
a=0.5542
b=3.051*10**-5

def Cp(T,a,b):    
    return a+b/T

def P_VDW(T,v):    
    return R*T/(v-b)-a/v**2

v=np.logspace(np.log10(b*1.1),np.log10(b*100),201)

plt.figure(dpi=300,figsize=(2,2))
plt.plot(v*10**6,P_VDW(300+273.15,v)/10**6)
plt.ylabel(r'$P$ $(MPa)$')
plt.xlabel(r'$v$ $(cm^3/mol)$')
plt.yscale('log')
plt.xscale('log')
```

#### let's solve the above expression for the VDW equation of state:

$\quad \Delta\underline{H}_{1\rightarrow2}^{R}=RT_2(Z_2-1)-RT_1(Z_1-1)+
\int_{\underline{V}_1}^{\underline{V}_2}\bigg[T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-P\bigg]dP$

first, we need to evaluate:

$\quad T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-P=\dfrac{a}{\underline{V}^2}$

integrating the above, we find that:

$\quad \Delta\underline{H}_{1\rightarrow2}^{R}=RT_2(Z_2-1)-RT_1(Z_1-1)-a\bigg(\dfrac{1}{\underline{V_{2}}}-\dfrac{1}{\underline{V_{1}}}\bigg)$

Adding back in the change for the ideal gas state, we find:

$\quad \Delta\underline{H}_{1\rightarrow2}=\Delta\underline{H}_{1\rightarrow2}^{R}+\Delta\underline{H}_{1\rightarrow2}^{IG}$

where

$\quad \Delta\underline{H}_{1\rightarrow2}^{IG}=\int_{T_1}^{T_2}a+bT dT=a(T_2-T_1)+\dfrac{b}{2}(T_2^2-T_1^2)$
