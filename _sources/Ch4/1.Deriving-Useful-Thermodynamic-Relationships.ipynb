# Deriving Useful Thermodynamic Properties

Up to this point in the class, we have used either the *ideal gas* properties or tabluated *real fluid* physical properties to compute the amount of heat and work required to change the state of substances, perform useful tasks like liquefaction, refrigeration, and power generation. In this chapter, we seek to develop a more generalized approach for such computations that allows for the estimation of *thermal* variables in terms of *volumetric* variables. 

The outline of the content in this section is as follows:

1. Methematical Preliminaries

2. Derivation of Maxwell's Equations

3. A summary of a few of the most useful expressions


## Mathematical Preliminaries

### Multivariate Calculus

Recall from calculus that if we have a function $\theta(X,Y,Z)$ that is a function of multiple variables, we can write the total differential for $\theta$ as:

$\quad d\theta=\bigg(\dfrac{\partial\theta}{\partial X}\bigg)_{Y,Z}dX+\bigg(\dfrac{\partial\theta}{\partial Y}\bigg)_{X,Z}dY+\bigg(\dfrac{\partial\theta}{\partial Z}\bigg)_{X,Y}dZ$

We will combine this mathematical fact with the combined form of the 1st and 2nd laws to derive useful quantities:

$\quad d\underline{U}=Td\underline{S}-Pd\underline{V}$

### Closed System Expressions: 

For closed system problems, we general compute the change in internal energy $\underline{U}$ in terms of other variables. 

Now lets see that in this combination of the first and second laws, there are differential expressions for $\underline{V}$ and $\underline{S}$ on the l.h.s. and $\underline{U}$ on the r.h.s. If we use our the calculus description above to write $\underline{U}(\underline{S},\underline{V})$, we can also write:

$\quad d\underline{U}=\bigg(\dfrac{\partial\underline{U}}{\partial \underline{V}}\bigg)_{\underline{S}}d\underline{V}+\bigg(\dfrac{\partial\underline{U}}{\partial \underline{S}}\bigg)_{\underline{V}}d\underline{S}$

Hopefully you can see where this is going, we can combine these two expressions to develop new thermodynamic relationships between $\underline{U}$, $\underline{V}$, and $\underline{S}$.

$\quad \bigg(\dfrac{\partial\underline{U}}{\partial \underline{V}}\bigg)_{\underline{S}}=-P$

$\quad \bigg(\dfrac{\partial\underline{U}}{\partial \underline{S}}\bigg)_{\underline{V}}=T$

We can go further, let's recall that the Helmholtz Free Energy is defined as: $A=U-TS$. Therefore, we can write:

$\quad d\underline{A}=d\underline{U}-T d\underline{S}-\underline{S} dT $

substituting for $d\underline{U}$, we get:

$\quad d\underline{A}=Td\underline{S}-Pd\underline{V}-T d\underline{S}-\underline{S} dT $

which simplifies to:

$\quad d\underline{A}=-Pd\underline{V}-\underline{S} dT $

Now we see that in this new combination of the first and second laws, there are differential expressions for $\underline{V}$ and $T$ on the r.h.s. and $\underline{A}$ on the l.h.s. If we use our the calculus description above to write $\underline{A}(\underline{V},T)$, we can also write:

$\quad d\underline{A}=\bigg(\dfrac{\partial\underline{A}}{\partial \underline{V}}\bigg)_{T}d\underline{V}+\bigg(\dfrac{\partial\underline{A}}{\partial T}\bigg)_{\underline{V}}dT$

now we can use our calculus to write:

$\quad \bigg(\dfrac{\partial\underline{A}}{\partial \underline{V}}\bigg)_{T}=-P$

$\quad \bigg(\dfrac{\partial\underline{A}}{\partial T}\bigg)_{\underline{V}}=\underline{S}$

### Open System Expressions:

For open system problems, we gernally compute the change in enthalpy $\underline{H}$ in terms of other state variables. As we did above we can transform the equation by employing the definition of the enthalpy $H=U+PV$.

$\quad d\underline{H}=d\underline{U}+Pd\underline{V}+\underline{V}dP$

again if we substitute for $d\underline{U}$, we get:

$\quad d\underline{H}=Td\underline{S}-Pd\underline{V}+Pd\underline{V}+\underline{V}dP$

or 

$\quad d\underline{H}=Td\underline{S}+\underline{V}dP$

Again we have now have $\underline{S}$ and $P$ on the r.h.s. and $\underline{H}$ on the l.h.s. If we use our the calculus description above to write $\underline{H}(\underline{S},P)$, we can also write:

$\quad d\underline{H}=\bigg(\dfrac{\partial\underline{H}}{\partial P}\bigg)_{\underline{S}}dP+\bigg(\dfrac{\partial\underline{H}}{\partial \underline{S}}\bigg)_{P}d\underline{S}$

and again we get:

$\quad \bigg(\dfrac{\partial\underline{H}}{\partial \underline{S}}\bigg)_T=T$

$\quad \bigg(\dfrac{\partial\underline{H}}{\partial P}\bigg)_{\underline{S}}=\underline{V}$

One more set of relationships can be obtained using the definition of the Gibb's Free Energy $G=H-TS$ or

$\quad d\underline{G}=d\underline{H}-Td\underline{S}-\underline{S}dT$

substituting for our differential expression for $d\underline{H}$, we get

$\quad d\underline{G}=Td\underline{S}+\underline{V}dP-Td\underline{S}-\underline{S}dT$

or 

$\quad d\underline{G}=\underline{V}dP-\underline{S}dT$

the theme continues, we again use calculus to recognize:


$\quad d\underline{G}=\bigg(\dfrac{\partial\underline{G}}{\partial P}\bigg)_{T}dP+
\bigg(\dfrac{\partial\underline{G}}{\partial T}\bigg)_{P}dT$

and now we get:

$\quad \bigg(\dfrac{\partial\underline{G}}{\partial T}\bigg)_P=-\underline{S}$

$\quad \bigg(\dfrac{\partial\underline{G}}{\partial P}\bigg)_{T}=\underline{V}$

## Maxwell's Equations

Alright so that was a useful exercise, we collected a few new ways to inter-relate properties, but we didn't really accomplish the goal we set out for ourselves - to be able to compute thermal variables in terms of the volumetric variables. To do so, we return to two of the expressions above useful for closed and open system problems:

$\quad d\underline{U}=Td\underline{S}-Pd\underline{V}$

$\quad d\underline{H}=Td\underline{S}+\underline{V}dP$

We can see that in the above, we have $d\underline{H}$ and $d\underline{U}$ in terms of $P$, $\underline{V}$, $T$ and $\underline{S}$, but we cannot calculate $d\underline{S}$

So let's use the approach explored above to develop expressions $\underline{S}\big(\underline{V},T\big)$ and $\underline{S}\big(P,T\big)$. 

### Expressions for Enthalpy and Internal Energy

Let's start with $\underline{S}\big(\underline{V},T\big)$:

$\quad d\underline{S}=\bigg(\dfrac{\partial\underline{S}}{\partial T}\bigg)_{\underline{V}}dT+
\bigg(\dfrac{\partial\underline{S}}{\partial \underline{V}}\bigg)_{T}d\underline{V}$

and $\underline{S}\big(P,T\big)$:

$\quad d\underline{S}=\bigg(\dfrac{\partial\underline{S}}{\partial T}\bigg)_{P}dT+
\bigg(\dfrac{\partial\underline{S}}{\partial P}\bigg)_{T}dP$

Ok. We are making some progress. If we could just replace the partial derivatives of $\underline{S}$ in the above, then we could complete the expression. For this, we are going to borrow two other tricks from calculus:

1. Chain Rule

2. Triple Product Rule

### Applying Chain Rule:


$\quad \bigg(\dfrac{\partial\underline{S}}{\partial T}\bigg)_{\underline{V}}=
\bigg(\dfrac{\partial\underline{S}}{\partial \underline{U}}\bigg)_{\underline{V}}\bigg(\dfrac{\partial\underline{U}}{\partial T}\bigg)_{\underline{V}}=\dfrac{1}{T}C_{V}$


$\quad \bigg(\dfrac{\partial\underline{S}}{\partial T}\bigg)_{P}=\bigg(\dfrac{\partial\underline{S}}{\partial \underline{H}}\bigg)_{P}\bigg(\dfrac{\partial\underline{H}}{\partial T}\bigg)_{P}=\dfrac{1}{T}C_P$


### Appling Triple Product Rule:

For differential expressions that obey:

$\quad d\underline{U}=\bigg(\dfrac{\partial\underline{U}}{\partial \underline{V}}\bigg)_{\underline{S}}d\underline{V}+
\bigg(\dfrac{\partial\underline{U}}{\partial \underline{S}}\bigg)_{\underline{V}}d\underline{S}$

The triple product rule states that 

$\quad \bigg(\dfrac{\partial\underline{U}}{\partial \underline{V}}\bigg)_{\underline{S}}=
-\bigg(\dfrac{\partial\underline{U}}{\partial \underline{S}}\bigg)_{\underline{V}}
\bigg(\dfrac{\partial\underline{S}}{\partial \underline{V}}\bigg)_{T}$

or 

$\quad \bigg(\dfrac{\partial\underline{S}}{\partial \underline{V}}\bigg)_{T}=
\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}$

and 


$\quad d\underline{H}=\bigg(\dfrac{\partial\underline{H}}{\partial P}\bigg)_{\underline{S}}dP
+\bigg(\dfrac{\partial\underline{H}}{\partial \underline{S}}\bigg)_{P}d\underline{S}$

The triple product rule states that 

$\quad \bigg(\dfrac{\partial\underline{H}}{\partial P}\bigg)_{\underline{S}}=
-\bigg(\dfrac{\partial\underline{H}}{\partial \underline{S}}\bigg)_{P}
\bigg(\dfrac{\partial\underline{S}}{\partial P}\bigg)_{T}$

or 

$\quad\bigg(\dfrac{\partial\underline{S}}{\partial P}\bigg)_{T}=
-\bigg(\dfrac{\partial \underline{V}}{\partial T}\bigg)_P$


The outcomes of the triple product rule applied to thermodynamic definitions are two of the equations called the **Maxwell Equations**

## Useful Expressions for Calculations

Recall, that we wanted an expression for $d\underline{S}$

$\quad d\underline{U}=Td\underline{S}-Pd\underline{V}$

$\quad d\underline{H}=Td\underline{S}+\underline{V}dP$


### Closed System Expressions

$\underline{S}\big(\underline{V},T\big)$:

$\quad d\underline{S}=\dfrac{C_V}{T}dT+
\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}d\underline{V}$

$\underline{U}\big(\underline{V},T\big)$:

$\quad d\underline{U}=C_{V}dT+
\bigg(T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-P\bigg)d\underline{V}$



### Open System Expressions

$\underline{S}\big(P,T\big)$:

$\quad d\underline{S}=\dfrac{C_P}{T}dT-
\bigg(\dfrac{\partial \underline{V}}{\partial T}\bigg)_PdP$

$\underline{H}\big(P,T\big)$:

$\quad d\underline{H}=C_{P}dT+
\bigg(\underline{V}-T\bigg(\dfrac{\partial \underline{V}}{\partial T}\bigg)_{P}\bigg)dP$


Note that these derivatives can be carried out on analytical forms of the equation of state



## Illustration

The implications of the above expressions are that if we have a volumetric equation of state and heat capacity information, we can directly calculate **thermal variables** with volumetric variables alone. Let's check our work first using the ideal gas law:

$\quad \dfrac{P\underline{V}}{RT}=1$

Plugging and chugging into the expressions above confirm our findings in Ch 3. that the change in internal energy and enthalpy of an ideal gas is only dependent on the heat capacities and the change in entropy is given logarithmic functions of the volumetric variables.

### Closed System Ideal Gas

$\underline{S}\big(\underline{V},T\big)$:

$\quad d\underline{S}=\dfrac{C_V}{T}dT+
\dfrac{R}{\underline{V}}d\underline{V}$

$\underline{U}\big(\underline{V},T\big)$:

$\quad d\underline{U}=C_{V}dT$



### Open System Expressions

$\underline{S}\big(P,T\big)$:

$\quad d\underline{S}=\dfrac{C_P}{T}dT-
\dfrac{R}{P}dP$

$\underline{H}\big(P,T\big)$:

$\quad d\underline{H}=C_{P}dT$
