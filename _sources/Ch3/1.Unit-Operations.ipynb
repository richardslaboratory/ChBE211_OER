# Unit Operations

The traditional role of a chemical engineer is to oversee the design, operation, and optimization of **unit operations** that are used to produce useful products. Unit Operations in physical plants are single peices of equipment that have a specific role in changing the state of matter and it is their combination that makes the conversion of one state of matter to another possible. Understanding the role of unit operations in process design is core to chemical engineering practice.

In this class, we will review three examples of useful processes that act on single component fluids that exist within a variety of industrial settings. These include:

- Power Generation

- Refrigeration

- Liquefaction

Before we can do that, we need to review common unit operations that are encountered in industrial settings that make these processes possible.

## Compressors

The schematic of an *adiabatic compressor* is shown below. A *well-designed* adiabatic compressor increases the pressure of a gaseous fluid **isoentropically**. A compressor is said to be well-designed if it fulfills the common assumptions below. 

Compressors work by using electric motors to turn high-speed rotors that cause a significant increase in the gas pressure and temperature. In a compressor care must be taken in its design to ensure that the temperature of the compressed gas does not exceed the operating temperature of the rotors, fan blades, bearings, and gaskets. 

<img src="images/Unit Operation - Compressor.png" width="300" >

Compressors are generally designed to achieve a desired internal compression ratio ($R$):

$$ R = \dfrac{P_2}{P_1} $$

where (1) is the inlet pressure and (2) is the outlet pressure. You will see in unit operations that this compression ratio defines the efficiency of the compression process, but also the design of the compressor. While outside the scope of this class, common compressors will include 

1. *adiabatic compression* - reversible compression where the gas is assumed to be compressed *adiabatically*

2. *isothermal compression* - reversible compression where cooling is performed during compression so that the compression is *isothermal*

2. *polytropic compression* - reversible compression where the fluid is neither *isothermal* nor *adiabatic*


### Common Assumptions for an Adiabatic Compressor:

- reversible operation 

- adiabatic (well-insulated/designed)

- operates at a steady state 

- Kinetic and potential energy changes are negligible

- Operates on a single gaseous phase - no condensation within a turbine is permissible

### Simplified Mass, Energy, and Entropy Balance Expressions:

1. Mass Balance Simplifies to: 

$$\dot{N}_1=\dot{N}_2$$

2. Energy Balance Simplifies to: 

$$\dot{W}_s=\dot{N}_1\big(\underline{H}_2-\underline{H}_1\big)$$

3. Entropy Balance Simplifies to:

$$\underline{S}_2=\underline{S}_1$$


## Turbines

The schematic of an *adiabatic turbine* is shown below. A *well-designed* adiabatic turbine extracts work **isoentropically** from the decrease in pressure of a high pressure gaseous fluid. 

Turbines work by using a high pressure gas to turn high speed rotors that convert mechanical energy to electricity and result in the reduction of pressure and temperature of the gas at the turbine outlet. In a turbine care must be taken to ensure that the working fluid and conditions of the expansion are such that the temperature of the gas does not drop below the critical temperature, risking condensation and damage of the high speed turbines.

<img src="images/Unit Operation - Turbine.png" width="300" >

Steam turbines are generally designed to maximize the work output from the available enthalpy in the incoming gas stream or:

$$ \dfrac{\dot{W}}{\dot{N}_1\underline{H}_2} $$

where (1) is the inlet and (2) is the outlet flow. You will see that this condition is satisfied when $\underline{H}_1 << \underline{H}_2$. 

For ideal gases, *gas turbines*, the work is maximized when inlet pressure and temperature are large and the outlet pressure and temperature is small (i.e., the enthaply differences between inlet and outlet are maximized). These high temperatures and pressures are achieved by combining in-line compression with combustion to heat and pressurize the air prior to expansion across the turbine. The [Department of Energy (DOE)](https://www.energy.gov/fecm/how-gas-turbine-power-plants-work) operates a funding program that has raised the operating temperature of metals within Turbines to exceed $800^{o} C$. Gas turbines are currently playing a critical role in the push to move away from coal-fired power plants.

For steam, *steam turbines*, the goal of the designer is similar to that of *gas turbines* with the caveat that significant condensation of the steam must be avoided within the turbine (steam quality must exceed 90% for the turbine to operate safely). Therefore, high inlet temperatures and pressures are desired. These pressures are achieved by optimizing the combustion within boilers where high pressure water is fed into a combustion chamber and vaporized to form super-heated steam at high pressures. Typical temperatures achievable in modern boilers exceed $600^{o}C$ and pressures approach $50$ bar.

An *adiabatic turbine* is said to be well-designed if it fulfills the common assumptions below. 


### Common Assumptions for an Adiabatic Compressor:

- reversible operation 

- adiabatic (well-insulated/designed)

- operates at a steady state 

- Kinetic and potential energy changes are negligible

- Operates on a single gaseous phase - no condensation within a turbine is permissible

### Simplified Mass, Energy, and Entropy Balance Expressions:

1. Mass Balance Simplifies to: 

$$\dot{N}_1=\dot{N}_2$$

2. Energy Balance Simplifies to: 

$$\dot{W}_s=\dot{N}_1\big(\underline{H}_2-\underline{H}_1\big)$$

3. Entropy Balance Simplifies to:

$$\underline{S}_2=\underline{S}_1$$


<div class="alert alert-block alert-warning">
<b>Note:</b> The design equations for the compressor and turbine are the same.
</div>


## Pumps

Pumps handle liquids and can be used for a variety of purposes including increasing the velocity, elevation, or the pressure of a liquid. In pumps, the density of the fluid remains unchanged due to the compression because fluids are incompressible. Therefore, the enthalpy change as a result of a pump is dominated by the pressure change of that fluid. 

<img src="images/Unit Operation - Pump.png" width="300" >

The efficiency of a pump is determined by how much pressure can be achieved at the outlet for a given amount of input work. While their design is complex and they are sensitive to inlet conditions, pumps routinely achieve pseudo-reversible pressure increases. A major challenge for pumps is avoiding *cavitation*, which is the formation of vacuum cavities in a continuous fluid. These vacuum *bubbles* can cause damage to the pump and efficiency losses. 

### Common Assumptions:

- reversible operation

- adiabatic (well-insulated/designed)

- operates at a steady state 

- Kinetic and potential energy changes are negligible - this will only be true in this class.

- Operates on a single liquid phase - no vaporization within a pump is permissible

### Simplified Mass, Energy, and Entropy Balance Expressions for Assumptions Above:

1. Mass Balance Simplifies to: 

$$\dot{N}_1=\dot{N}_2$$

2. Energy Balance Simplifies to: 

$$\dot{W}_s=\dot{N}_1\big(\underline{H}_2-\underline{H}_1\big)\approx \dot{N}\underline{V}\big(P_2-P_1\big)$$

3. Entropy Balance Simplifies to:

$$\underline{S}_2=\underline{S}_1$$

<div class="alert alert-block alert-warning">
<b>Note:</b> The design equation for pumps is almost identical to that of turbines and compressors but with the enthalpy change for the liquid depending mainly on the difference in pressure from the outlet to the inlet due to the fluids incompressible nature. 
</div>



## Throttle Valves

Throttle valves are commonly used to rapidly and *isoenthalpically* decrease the temperature and pressure of a high pressure vapor. 

<img src="images/Unit Operation - Throttle Valve.png" width="400" >

Throttle valves have a relatively simple operation and in this class they are commonly used to perform *liquefaction* or making a saturated liquid. This is useful for a variety of purposes and we will see throttle valves appear both in refrigerators and in liquefaction applications.

### Common Assumptions:

- adiabatic (well-insulated/designed)

- operates at a steady state 

- Kinetic and potential energy changes are negligible - this will only be true in this class.

- Operates on a single liquid phase - no vaporization within a pump is permissible

- Connected to a flash drum where the condensate is collected

### Simplified Mass, Energy, and Entropy Balance Expressions for Assumptions Above:

1. Mass Balance Simplifies to: 

$$\dot{N}_1=\dot{N}_2$$

2. Energy Balance Simplifies to: 

$$\underline{H}_2=\underline{H}_1$$

3. Entropy Balance Simplifies to:

$$0=\dot{N}_1\big(\underline{S}_2-\underline{S}_1\big)+\dot{S}_{gen}$$

<div class="alert alert-block alert-warning">
<b>Note:</b> The design equations for a throttle valve emphasize that the process is isoenthalpic and one can anticipate large amounts of entropy generation!
</div>

## Heat Exchangers

Heat exhangers use the thermal contact between two fluids/gases to change their temperature. This can be for both heating and cooling applications, but the general process is schematically represented below. 

<img src="images/Unit Operation - Heat Exchanger.png" width="400" >

The design of heat exchangers is a particularly important area of chemical engineering and we will only mention their role here in the context of liquefaction. You will learn much more in *Heat Transfer*

### Common Assumptions:

- operates at a steady state 

- Kinetic and potential energy changes are negligible - this will only be true in this class.

- Operates on a two single phase fluids - generally without phase change, but this is not always the case


### Simplified Mass, Energy, and Entropy Balance Expressions for Assumptions Above:

1. Mass Balance Simplifies to: 

$$\dot{N}_1=\dot{N}_2; \dot{N}_3=\dot{N}_4$$

2. Energy Balance Simplifies to: 

$$\dot{N}_3\big(\underline{H}_4-\underline{H}_3\big)=\dot{N}_1\big(\underline{H}_2-\underline{H}_1\big)$$

3. Entropy Balance Simplifies to:

$$0=\dot{N}_3\big(\underline{S}_4-\underline{S}_3\big)+\dot{N}_1\big(\underline{S}_2-\underline{S}_1\big)+\dfrac{\dot{Q}}{T_b}+\dot{S}_{gen}$$

In the above $T_b$ is the bulk temperature of the heat exchanged between the two fluids during their interaction.


## Boilers

Boilers use heat exchangers to boil a liquid fed into a flash drum. Boilers operate *isobarically* with the phase change occuring within the system and the output is typically a supersaturated vapor. 

<img src="images/Unit Operation - Boiler.png" width="400" >

The goal of boilers is to fully utilize all the input thermal energy from the heat exchanger to transform a liquid to a vapor with a desired outlet temperature and pressure. 

### Common Assumptions:

- operates at a steady state 

- Kinetic and potential energy changes are negligible - this will only be true in this class.

- *isobaric operation* within phase change within the system


### Simplified Mass, Energy, and Entropy Balance Expressions for Assumptions Above:

1. Mass Balance Simplifies to: 

$$\dot{N}_1=\dot{N}_2$$

2. Energy Balance Simplifies to: 

$$\dot{Q}=\dot{N}_1\big(\underline{H}_2-\underline{H}_1\big)$$

3. Entropy Balance Simplifies to:

$$0=\dot{N}_1\big(\underline{S}_2-\underline{S}_1\big)+\dfrac{\dot{Q}}{T_b}+\dot{S}_{gen}$$

In the above $T_b$ is the bulk temperature of the heat exchanged between the two fluids during their interaction.

## Condenser

Condensers use coolant to condense low pressure vapor fed into the inlet. Condensers operate *isobarically* with the phase change occuring within the system and the output is typically a saturated liquid. 

<img src="images/Unit Operation - Condenser.png" width="400" >

The goal of condensers is to efficiently extract heat from the vapor so that it can be recycled back into a closed process. Condensers typically operate using cooling water from natural streams/rivers and in larger scale power plants the heat the water gains must be exhausted to the atmosphere prior to the water being returned to the environment. To accomplish this, often the cooling water is fed into a cooling tower which sprays hot water droplets, causing them to fall. During their fall, a small fraction of the water evaporates cooling the remainder of the water in the tower to permissible temperatures.

### Common Assumptions:

- operates at a steady state 

- Kinetic and potential energy changes are negligible - this will only be true in this class.

- *isobaric operation* within phase change occuring from a vapor to a liquid phase.


### Simplified Mass, Energy, and Entropy Balance Expressions for Assumptions Above:

1. Mass Balance Simplifies to: 

$$\dot{N}_1=\dot{N}_2$$

2. Energy Balance Simplifies to: 

$$\dot{Q}=\dot{N}_1\big(\underline{H}_2-\underline{H}_1\big)$$

3. Entropy Balance Simplifies to:

$$0=\dot{N}_1\big(\underline{S}_2-\underline{S}_1\big)+\dfrac{\dot{Q}}{T_b}+\dot{S}_{gen}$$

In the above $T_b$ is the bulk temperature of the heat transfered to the fluid within the boiler.

</div>
