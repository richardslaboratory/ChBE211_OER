While there are many examples of heat engines, in thermodynamics for chemical engineers, we are principally concerned with heat engines that utilize fluids to convert heat to mechanical work.


A Rankine cycle is the most common way to produce electrical energy in the United States. Generally, coal, biofuels, concentrated photothermal energy, and nuclear power is used to produce a high pressure fluid (often steam) which is then passed through a turbine, condensed, and recycled back into the process. #### Figure of Merit

The efficiency of a Rankine Power Cycle is given by the efficiency of any heat engine:

$$\eta=-\dfrac{\dot{W}}{\dot{Q}_H}
=\dfrac{\dot{W}_{1-2}+\dot{W}_{3-4}}{\dot{Q}_{2-3}}
=\dfrac{(\underline{H}_4-\underline{H}_{3})-\underline{V_1}(P_2-P_1)}{(\underline{H}_3-\underline{H}_{2})}$$

<div class="alert alert-block alert-info">
<b>Note:</b> the above equations assume the pump and turbine operate reversibly.
</div>

Modern boilers can handle up to 30 MPa and maximum operating temperatures of 650 C. Given the condensation takes place below 100 C, the Carnot Efficiency can be calculated as follows:

$$\eta_{C}=1-\dfrac{T_C}{T_H}=1-\dfrac{373.15}{923.15}=59.6\%$$

The real efficiency is much lower than this Carnot Efficiency. Lets see why below using some basic assumptions:

1. Assume the water entering the boiler is at 100 C and 30 MPa.
2. Assume that it exits the boiler at 650 C
3. Assume that steam exits the turbine at 0.1 MPa 

Then the state points can be summarized in tabular form as:

| state | $T\:(^o C)$ | $P\:(MPa)$ | $\underline{H}\:(kJ/mol)$  | $\underline{S}\:(J/mol\cdot K)$ | 
| :-: | :-:   | :-:                   | :-:       | :-:         | 
| 1 | 100 | 0.1 | 48.19  | 24.5 |
| 1 | 100 | 30 | 7.95 | 24.5 |
| 3 | 650 | 30 | 64.84 | 115.43 |
| 4 | 100 | 0.1 | 42.08 | 115.43 | |

if we also identify the molar volume of steam, $\underline{V}_1\approx\underline{V}_2=0.0180\:L/mol$, then we can calculate the efficiency as:

$$\eta=\frac{(42.08-64.84)0.018(30-0.1)}{7.95-64.84}=21.5\%$$



## Piston Based Heat Engines

In heat engines that derive useful work from the action of a moving boundary (i.e., pistons), we should appreciate that at a thermodynamic cycle must be completed that turns the action of compression and expansion of the cylinder volume to changes in internal energy of the gas within that volume. Based on this understanding at each point throughout the cycle state of the gas will be determined determined by the relative amounts of heat added and removed from the cylinder, the pathway taken between each state point, and the volume change undertaken by the cylinder during expansion.

If the expansion and compression cycles are reversible, then a mass, energy, and entropy balance around the closed volume  at any step throughout the cycle reveals the following equation:

1. $dN=0$
2. $dU=\dot{Q}-PdV$
3. $dS=\dfrac{\dot{Q}}{T}$

Combining 1 and 3 by equating the heat terms, we can see that:

$\quad dA=dU-TdS=-PdV$

In other words, the free energy change associated with expansion and contraction is described by the **Helmholtz Free Energy**, which represents the energy available for to do useful work. Because of the sine of the right hand side of the equation above, we can appreciate that:

- if $dV>0$, then $dA<0$ and the free energy of the gas has been decreased. In other words, expansion reduces the capacity of the gas to do useful work

- if $dV<0$, then $dA>0$ and the free energ yof the gas has been icnreased. In other words, compression increases the capacity of the gas to do useful work.

The details of different closed system cycles will depend on the path used to effect a chance of state of the gas.

- reversible, adiabatic expansion/compression is always **isoentropic**, (i.e., $dS=0$)

- **isothermal** expansion/compression of an ideal gas is always at constant internal energy (i.e., $dU=0$)

- **isobaric** expansion/compression are those a the volume change is affected through the exchange of heat where the piston undergoes expansion/compression without exerting a net force on the surroundings.

## Carnot Cycle

A Carnot cycle is a conceptual model to understand the relationship linking work and heat flow that does boundary work. This is conceptually the most basic cycle that can be constructed to do work based on pistons. 

Consider the following:

- N moles of an ideal gas into a cylinder
- Expansion of the cylinder and compression of the cylinder occurs reversibly
- The cylinder is well-insulated from its environment except when its head is in contact with a thermal reservoir
- The shaft is connected to a motor from which we wish to extract work
- The head can be contacted with a hot reservoir and a cold reservoir maintained at a constant temperature $T_H$ and $T_C$

Based on the following facts, we can envision designing a cycle that generates net work via the following **4 step** pathway.

Insert Drawing Here

### The Pathways defined by the Carnot Cycle are listed below.

| Step   | Description      | Pathway                | Heat                    | Work |
| :-:    | :-:              | :-:                    | :-:                     | :-:  |
| 1-2  | Cold Reservoir   | isothermal compression | $\dot{Q}_{C}$ | $-PdV$ |
| 2-3  | Recovery Stroke  | adiabatic  compression | $0$           | $-PdV$         |
| 3-4  | Hot Reservoir    | isothermal expansion   | $\dot{Q}_{H}$ | $-PdV$ |
| 4-1  | Power Stroke     | adiabatic  expansion   | $0$           | $-PdV$         |

### Based on these Paths - We can define the State of System
Based on the paths above, we can assemble a property table assuming the working fluid is an ideal gas:

| Step   | Pathway        | $T$ | $P$ | $\underline{V}$ | $\underline{S}$ | $\underline{U}$ |
| :-:    | :-:                | :-: | :-: | :-:             | :-:             | :-:             |
| 1      | | $T_C$ | $P_{1}$ | $\underline{V}_1$  | $0$ | $0$ |
| 1-2    | Isothermal |  | |  |  |  | 
| 2      | | $T_C$|  $P_2=K P_{1}$ | $\dfrac{RT_C}{P_2}$  | $-R\ln{K}$ | $0$ | 
| 2-3    | Isoentropic |  | |  |  |  |
| 3      | | $T_H$ | $P_3=P_2\bigg(\dfrac{T_H}{T_C}\bigg)^{C_{p}^{IG}/R}$ | $\dfrac{RT_H}{P_3}$  | $-R\ln{K}$ | $C_v^{IG}(T_H-T_C)$ | 
| 3-4    | Isothermal |  | |  |  |  |
| 4      | | $T_H$ | $P_4=P_1\bigg(\dfrac{T_H}{T_C}\bigg)^{C_{p}^{IG}/R}$ | $\dfrac{RT_H}{P_4}$  | $0$ | $C_v^{IG}(T_H-T_C)$ | 
| 4-1    | Isoentropic |  | |  |  |  | 

### In the table above, we have specified the following properties

- The temperature in the hot reservoir, $T_H$

- The temperature in the cold reservoir, $T_C$

- The compression ratio, $K=P_2/P_1$

- The initial pressure in the cylinder, $P_1$

- The reference temperature $T_R=T_C$ and reference entropy is $\underline{S}_1$

- The ideal gas heat capacity is independent of temperature and given by $C_v^{IG}$

#### Other pieces of information we have used are

- Equation of state: 

$$ P\underline{V}=RT $$

- Entropy Change of Ideal Gas: 

$$dS=\dfrac{C_{v}^{IG}}{T}dT+\dfrac{R}{\underline{V}}d\underline{V}$$

$$dS=\dfrac{C_{p}^{IG}}{T}dT-\dfrac{R}{P}d\underline{P}$$

Based on the above, for an *isoentropic* property change, we can say that:

$$\bigg(\dfrac{T_f}{T_i}\bigg)^{C_{p}^{IG}/R}=\dfrac{P_f}{P_i}$$

### Lets plot this:

```python
import numpy as np
import matplotlib.pyplot  as plt

R=8.314 #J/mol*K

##returns the pressure and volume of a carnot cycle for an IG, based on standard properties
def Carnot_IG(Th=800,Tc=298,P1=1,Cv=28,K=10):
    '''
    Input:
    
    Th is the hot reservoir temperature in K
    Tc is the cold reservoir temperature in K
    P1 is the minimum pressure throughout the cycle in bar
    Cv is the constant volume heat capacity in J/molK
    K=P2/P1 is the compression ratio and is unitless
    
    Output:
    PV object is a list of tuple pairs containing PV values at each state point
    '''
    Cp=Cv+R
    ##State point one
    P1=P1*10**5 ##unit conversion from bar to Pa
    V1=R*Tc/P1
    T1=Tc
    S1=0
    U1=0
    A1=U1-T1*S1
    
    #State point two
    P2=K*P1 ##P2 calculation
    V2=R*Tc/(P2) ##V2 calculation
    T2=Tc
    S2=-R*np.log(K)
    U2=0
    A2=U2-T2*S2
    
    #State point three
    P3=P2*(Th/Tc)**(Cp/R)
    V3=R*Th/(P3)
    T3=Th
    S3=S2
    U3=(Th-Tc)*Cv
    A3=U3-T3*S3
    
    ##State point four
    P4=P1*(Th/Tc)**(Cp/R)
    V4=R*Th/(P4)
    T4=Th
    S4=0
    U4=(Th-Tc)*Cv
    A4=U4-T4*S4
    
    
    ##construct the plottables 
    V_=np.asarray([V1,V2,V3,V4,V1])*10**4##convert to cm^3/mol
    P_=np.asarray([P1,P2,P3,P4,P1])/10**5##convert to bar
    T_=np.asarray([T1,T2,T3,T4,T1])-273.15##conver to C
    S_=np.asarray([S1,S2,S3,S4,S1])
    U_=np.asarray([U1,U2,U3,U4,U1])
    A_=np.asarray([A1,A2,A3,A4,A1])
    
    ##PV Pathways
    path1=[np.logspace(np.log10(V1),np.log10(V2)),
           R*Tc/np.logspace(np.log10(V1),np.log10(V2))] ##isothermal pathway
           
    path3=[np.logspace(np.log10(V3),np.log10(V4)),
           R*Th/np.logspace(np.log10(V3),np.log10(V4))] ##isothermal pathway
    
    path2=[np.logspace(np.log10(V2),np.log10(V3)),
           P2*(V2/np.logspace(np.log10(V2),np.log10(V3)))**(Cp/Cv)] ##isoentropic pathway
    
    path4=[np.logspace(np.log10(V4),np.log10(V1)),
           P4*(V4/np.logspace(np.log10(V4),np.log10(V1)))**(Cp/Cv)] ##isoentropic pathway
    
    PVpath_=[path1,path2,path3,path4]
    
    ##TS Pathways
    path1=[np.linspace(T1,T2),np.linspace(S1,S2)]
    path2=[np.linspace(T2,T3),np.linspace(S2,S3)]
    path3=[np.linspace(T3,T4),np.linspace(S3,S4)]
    path4=[np.linspace(T4,T1),np.linspace(S4,S1)]
    
    TSpath_=[path1,path2,path3,path4]
    
    return V_,P_,T_,S_,U_,A_,PVpath_,TSpath_

V_,P_,T_,S_,U_,A_,PVpath_,TSpath_=Carnot_IG(Th=900,Cv=12,K=10)

fig,ax=plt.subplots(dpi=300,figsize=(5,2),ncols=2)
fig.subplots_adjust(wspace=0.5)


##plot the state points
ax[0].plot(V_,P_,marker='s',linestyle='')
c_=['b','k','r','k']
for i,path in enumerate(PVpath_):
    v=path[0]*10**4
    p=path[1]/10**5
    ax[0].plot(v,p,c=c_[i])

##plot the state points
ax[1].plot(S_,T_,marker='s',linestyle='')

for i,path in enumerate(TSpath_):
    T=path[0]-273.15
    S=path[1]
    
    ax[1].plot(S,T,c=c_[i])  

ax[0].set_yscale('log')
ax[0].set_xscale('log')
ax[0].set_ylabel(r'$P$ $(bar)$')
ax[0].set_xlabel(r'$V$ $(cm^3/mol)$')

ax[1].set_ylabel(r'$T$ $(^{o}C)$')
ax[1].set_xlabel(r'$S$ $(J/(mol\cdot K)$')
```
