# Conservation of Energy and the 1st Law of Thermodynamics

## First Law of Thermodynamics - Equivalence of work, heat, and energy

The statement of the *conservation of energy* is the **1st Law of Thermodynamics**. Clausis put it this way in regards to a thermodynamic cycle:


### Clausius Definition of the First Law

"In all cases in which work is produced by the agency of heat, a quantity of heat is consumed which is proportional to the work done; and conversely, by the expenditure of an equal quantity of work an equal quantity of heat is produced"

This statement has two critical elements to it.

1. The fact that work is produced

2. The fact that heat is consumed

The interplay between work, $W$, and heat, $Q$, is a critical topic in thermodynamics.

We are primarily concerned with how work and heat are used to change the state of fluids in this class. For example, the addition of heat to water to produce steam can be used to drive a turbine to produce work. In this example the heat has the effect of changing the energy of fluid water which can be used to transfer that thermal internal energy to mechanical energy during the steam's expansion through the turbine.

### More General description
Another way to frame the statement of the first law of thermodynamics is that any change in energy, $\Delta E$, of any system is given by the amount of heat, $Q$, added/removed and the amount of work, $W$ , done/by that system. 

(1) $\Delta E=Q + W $

### Types of Energy 
There are three important forms of energy that contribute to $\Delta E$ that can be exploited to transform thermal energy to work and visa versa for useful purposes. 

1. **Kinetic Energy** - This is the energy associated with mass moving in an inertial reference frame, $E_{KE}=0.5mv^2$ 
    - $v$ is the velocity of the object


2. **Potential Energy** - This is the energy associated with an object because of its position relative to other objects. One example of potential energy is that which arises due to gravity, $E_{PE}=mgh$.   
    - $g$ is the acceleration that the body experiences in a gravitation field
    - $h$ is the distance separating the objects.
 
<div class="alert alert-block alert-info">
<b>Note:</b> There are many other types of potential energy that arise from other fields including magnetic, electric, and others.
</div>
 

3. **Internal Energy** - This is the energy associated with the thermal energy contained within a material's molecular bonds and its contribution of energy is given as $E_{IE}=U$
    
    - The internal energy of a dilute monoatomic gas is a function of the kinetic energy of all its atoms, i.e. its temperature
    
    - We write this  where $U$ is the internal energy

We will see in this class that it is the role of unit operations to efficiently perform these conversions and the role of you as an engineer to design those unit operations to perform the conversion of heat to work as efficiently as possible.


## Conservation of Energy

The full statement for the *conservation of energy* can be derived in a similar fashion as the mass balance expression.

To start, we will set $\theta=E$ or the energy of a system. From above, we know that the energy can be given as the sum of the kinetic, potential, and internal energy of the system, $E=E_{KE}+E_{PE}+U$. If we neglect the kinetic energy and potential energy contributions, or $E_{KE}+E_{PE}<<U$, then we can say that $\theta \approx U$ and we can then write the energy balance expression as:


$$\dfrac{dU}{dt}=\sum_{k} \dot{U}_k+\sum_{j} \dot{Q}_j+\sum_{l} \dot{W}_l$$

where 

- $\dot{U}_k$ are the energy associated with the mass flows across the system boundary

- $\dot{Q}_j$ are the energy associated with the heat flows across the system boundary

- $\dot{W}_l$ are the energy due to the work done on or by the system

- recall that on the L.H.S. $\dfrac{dU}{dt}$ is the rate of change of energy of within the system due to mass, energy, and work flows into and out of the system

- recall that $\dot{U}_{gen}=0$ because energy is a *conserved* quantity

### Definition of Work:

In order to simplify things a little bit it is useful to distinguish between a couple different kinds of work that can be done. These include:

- Shaft Work

- Boundary Work

- PV Work

#### Shaft work: 

Shaft work, $\dot{W}_s$, is electrical or mechanical work that is done by or on the surroundings (external to the system)

#### Boundary work: 

Boundary work, $\dot{W}_b$, is the work done due to changes in the system boundaries. In some cases, the system's volume is not constant (Ex. filling a balloon). In these cases, the expansion/contraction of the boundary leads to work being done by/on the system. In the differential form: $\dot{W}_b=-P_{ext}\dfrac{dV}{dt}$

- In this equation, $P_{ext}$ is the pressure of surroundings acting on the boundary that we assume to be uniform.

- The minus sign guarantees that:

    - Expansion of the boundary:  is work being done by the system on its surroundings 
    - Compression of the boundary:  is work being done by the surrounds on the system.

## Definition of Enthalpy:

In thermodynamics, we call the energy associated with the flow of mass the enthalpy, $H$, and it is defined as $H=U+PV$. Therefore Eq (2) can be further written as:

$$\dfrac{dU}{dt}=\sum_{k} \big(\dot{H}\big)_k+\sum_{j} \dot{Q}_j+ \dot{W}_s + \dot{W}_b $$

Finally, acknowledging that $U$ and $\dot{H}$ given here are extrinsic quantities, we can write the equation equivalently in its intrinsic form. 

$$\dfrac{d(N\underline{U})}{dt}=\sum_{k} \big(\dot{N}\underline{H}\big)_k+\sum_{j} \dot{Q}_j+ \dot{W}_s + \dot{W}_b$$ 

- it is important to understand that on the left hand side, the $N$ remains inside the time derivative as the mass within the system can change with time. This form will be most useful for **closed system problems**


## The *Conservation of Energy*

$$\dfrac{dU}{dt}=\sum_{k} \big(\dot{N}\underline{H}\big)_k+\sum_{j} \dot{Q}_j+ \dot{W}$$

- This last form is the most common form of the *conservation of energy* and it is the starting point for solving problems.
- In this form, the sum of shaft and boundary work is implied such that $\dot{W}=\dot{W}_s+\dot{W}_b$
