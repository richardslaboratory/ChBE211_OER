# Changes of State 


The **state** of a system is described by its **state properties**. In thermodynamics, we generally consider a system to be defined by a region of space that is separated from its surroundings by a boundary. We have learned that mass and energy can flow across that boundary to change the state of the system and internal gradients within the system can result in state changes via the accumulation of entropy.

The generation is of central importance to the study of thermodynamics, though in this class we focus mainly on reversible processes (i.e. processes where $\dot{S}_{gen}=0$). In practice all state changes effected by exchange of mass and energy lead to $\dot{S}_{gen}>0$ as a direct consequence of the formation of concentration, thermal, and velocity gradients internal to the system the lead to dissipation. The direct effect is that irreversible processes produce less heat and require more energy than their equivelant *reversible* counterpart.

A **central assumption** of state changes is that the change of state is *path independent*. Meaning for any physical process, if we desire to describe the change of state from point $A$ to point $B$, we can choose the path that is most convenient for computation and not be constrained to the path or device being analyzed. This **central fact** allows us to simplify calculations in two important ways.

1. To evaluate changes in state, we can choose a path such that:

    - There is no shaft work
    
    - There is only one mass flow
    
    - The mass and heat flows at a common temperature
    
    - The state change occurs reversibly

    
2. Using (1), we can use the change of state from $A$ to $B$ to compute the heat and mass flows for the real process. While state variables are path independent, the heat and mass flows are not and must be described for the path or device being analyzed

What follows is the derivation of the equation that describes the change in state of a **closed** and **open** system that adheres to the criteria in (1). We will see below the power of this approach at simplifying computations.

## Differential entropy change of state

Subject to the constraints of (1), we can write the mass energy and entropy balance expressions as follows:

(1) $\dfrac{dM}{dt}=\dot{M}$


(2) $\dfrac{dU}{dt}=\dot{M}\hat{H}+\dot{Q}-P\dfrac{dV}{dt}$


(3) $\dfrac{dS}{dt}=\dot{M}\hat{S}+\dfrac{\dot{Q}}{T}$

If we acknowledge that the change in state from point $A$ to $B$ occurs over a finite interval of time $dt$, then the equations above can be rewritten as:

(1) $dM=\dot{M}dt$


(2) $dU=\dot{M}dt\hat{H}+\dot{Q}dt-PdV$


(3) $dS=\dot{M}dt\hat{S}+\dfrac{\dot{Q}dt}{T}$

Combining the three equations is accomplished by substituting (1) for $\dot{M}dt=dM$ in (2) and (3), and $\dot{Q}dt=TdS-\dot{M}dtT\hat{S}$. The result is:

(2) $dU=dM\hat{H}+TdS-dM T\hat{S}-PdV$

Recognizing the common $dM$ term above, we can employ the definition of the Gibb's Free Energy of a System $G=H-TS$ to regroup terms and simplify the expression further

(2) $dU=TdS-PdV + \hat{G}dM$

What has this accomplished? Well if we examine the form of this equation, we can describe changes in internal energy and volume of a system in terms of changes in entropy and mass in a way that depends only on the path chosen.

This equation simplifies further for a *closed system* as:

$dU=TdS-PdV$

which on a molar basis is written as:

$d\underline{U}=Td\underline{S}-Pd\underline{V}$

This final form is powerful. We can use it to calculate changes in state even for systems where the flow is open.

## Illustration

An ideal gas with a constant heat capacity $C_p=29.3$ $J/(mol\cdot K)$ is compressed in a well-designed compressor. The flow rate is $2.5$ $mol/s$ and the input temperature and pressure are $1$ $bar$ and $290$ $K$ respectively. The output pressure is $10$ $bar$. Solve for the work and the temperature of the output stream.

### Simplify the mass, energy, and entropy balance expression 

As we have seen before, simplification of the mass, energy, and entropy balance expressions for this scenario yields:

- From the mass + energy balance, we get: $\dot{W}=\dot{N}\big(\underline{H}_2-\underline{H}_1\big)$ - note we have assumed steady state, reversibility, and adiabatic.

- From the mass + entropy balance, we get: $\underline{S_2}=\underline{S_1}$

### Identify the state information necessary to solve the problem

For an ideal gas, we know that we can use the relationship $\underline{H}(T)-\underline{H}(T_r)=C_P\big(T-T_R\big)$ to show that $\dot{W}=\dot{N}C_P \big(T-T_R\big)$. Therefore, we need to find $T_2$ to solve this equation.


### Solve for uknown temperature at outlet.

In order to find $T_2$, we will use the expression, 

$d\underline{U}=Td\underline{S}-Pd\underline{V}$.

Then we substutite using our known gas properties $d\underline{U}=C_V dT$ to arrive at

$C_V dT=Td\underline{S}-Pd\underline{V}$.

Since both the pressue and temperature of the gas are changing due to the compression but the entropy is not, it is convenient to isolate the $d\underline{S}$ term on the R.H.S, which should be 0. We do this by dividing both sides of the equation by $T$ and rearraging to yiled

$\dfrac{C_V}{T}dT+\dfrac{P}{T}d\underline{V}=0$.

Finally, we can use our known ideal gas properties to write that $P=\dfrac{RT}{\underline{V}}$ to further simplify as

$\dfrac{C_V}{T}dT+\dfrac{R}{\underline{V}}d\underline{V}=0$.

The above expression can be directly integrated to obtain the following expression:

$C_V ln\bigg(\dfrac{T_2}{T_1}\bigg)+R ln\bigg(\dfrac{\underline{V}_2}{\underline{V}_1}\bigg)=0$

or using the properties of logs, we find:

$\bigg(\dfrac{T_2}{T_1}\bigg)^{C_V/R}=\dfrac{\underline{V}_1}{\underline{V}_2}$

At this point, we can use the gas law to substitute for $\underline{V}=\dfrac{RT}{P}$ and write

$\bigg(\dfrac{T_2}{T_1}\bigg)^{C_V/R}=\dfrac{T_1 P_2}{T_2 P_1}$

or 

$\bigg(\dfrac{T_2}{T_1}\bigg)^{C_V/R+1}=\dfrac{P_2}{P_1}$

Finally, we have an expression we can use to find the temperature $T_2=554 K$.

### Finding the work

The work can be easily computed now that the temperature is known and $\dot{W}=19.6 kW$

### Key Insight from this Illustration

The elegant approach to solving this problem recognized that a change in state is **path independent**, and therefore, we can select the path that is most convenient to compute the property change. In this case, we used the change in internal energy of an ideal gas at constant entropy to compute the ratio between the pressures and the temperatures. We then utilized the properties that we determined to plug back into our mass, energy, and entropy balance to solve. The path independence of thermodynamic state variables is one of most powerful concepts in this class.
