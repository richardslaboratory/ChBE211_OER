# Expansion of a Gas

## Applications of the Mass and Energy Balance Expression

When a high pressure gas is expanded, generally its temperature and pressure decreases. This expansion can be used for useful purposes. In this class, we will see three scenarios:

1. [Steady state, adiabatic expansion of a gas across a valve](#1.1)

2. [Steady state, adiabatic expansion of a gas across a turbine](#1.2)

3. [Transient, adiabatic expansion of a gas within a closed piston](#1.3)

Last lecture, we derived the macroscopic mass and energy balance expressions given by (1) and (2) below:

$\quad(1)\quad\dfrac{dN}{dt}=\sum_{k} \big(\dot{N})_k$

$\quad(2)\quad\dfrac{dU}{dt}=\sum_{k}\big(\dot{N}\underline{H}\big)_k+\dot{Q}+ \dot{W}$

Today, we will simplify the balance expressions for the three scenarios above and see the results. Before we do, we first need to introduce new definitions:

1. **Steady state** - steady state means that there is no accumulation or depletion within the control volume. The implication is:

$$\dfrac{d}{dt}=0$$

<div class="alert alert-block alert-info">
<b>Note:</b> The opposite of a steady state system is a transient one, where $\dfrac{d}{dt}\ne 0$.
</div>


2. **Adiabatic** - adiabatic processes are those that are well insulated such that no heat can enter/exit the system across the control volume. The implication is:

$$\dot{Q}=0$$

3. **Closed** - closed systems are those that do not allow mass to flow across the system boundary. The implication of this statement is:

$$\dot{M}=0$$


<div class="alert alert-block alert-info">
<b>Note:</b> TThe opposite of a closed system is an open system, where $\dot{M}\ne0$.
</div>

Armed with these assumptions, we can simplify the mass and energy balance expressions for the three cases listed above.

<a id="1.1"></a>
### Steady state, adiabatic expansion of a gas across a valve

#### (1) Mass Balance Expression

$\dfrac{dN}{dt}=\sum_{k} \big(\dot{N})_k$

- steady state imples $\dfrac{d}{dt}=0$, so $\dfrac{dN}{dt}=0$
- there is a single input stream (1) and a single output stream (2) such that $\sum_k\big(\dot{N}\big)_k=\dot{N}_1+\dot{N}_2$

The **Conservation of Mass** thus simplifies to:

$$0=\dot{N}_1+\dot{N}_2$$ 

or 

$$\dot{N}_1=-\dot{N}_2$$

<div class="alert alert-block alert-info">
<b>Note:</b> the sine convention for mass entering and leaving the system is preserved by the conservation of mass.
</div>

#### (2) Energy Balance Expression
$\dfrac{dU}{dt}=\sum_{k}\big(\dot{N}\underline{H}\big)_k+\dot{Q}+ \dot{W}$

- steady state imples $\dfrac{d}{dt}=0$, so $\dfrac{dU}{dt}=0$
- there is a single input stream (1) and a single output stream (2) such that $\sum_k\big(\dot{N}\underline{H}\big)_k=\dot{N}_1\underline{H}_1+\dot{N}_2\underline{H}_2$
- the expansion is adiabatic such that $\dot{Q}=0$
- there is neither shaft nor boundary work such that $\dot{W}=0$

The **Conservation of Energy** thus simplifies to:

$$0=\dot{N}_1\underline{H}_1+\dot{N}_2\underline{H}_2$$


#### (1+2) Combined Expressions

We can now combine the above expressions by setting $\dot{N}_1=-\dot{N}_2$ from (1) and subsituting into (2)

$$0=\dot{N}_1\underline{H}_1-\dot{N}_1\underline{H}_2$$

or 

$$\underline{H}_2=\underline{H}_1$$

We conclude from this exercise that the steady state, adiabatic expansion across a valve is **isoenthalpic**

<a id="1.2"></a>
### Steady state, adiabatic expansion of a gas across a turbine

This derivation proceeds as above, except now there is shaft work

#### (1) Mass Balance Expression

$$\dfrac{dN}{dt}=\sum_{k} \big(\dot{N})_k$$

- steady state imples $\dfrac{d}{dt}=0$, so $\dfrac{dN}{dt}=0$
- there is a single input stream (1) and a single output stream (2) such that $\sum_k\big(\dot{N}\big)_k=\dot{N}_1+\dot{N}_2$

The **Conservation of Mass** thus simplifies to:

$$0=\dot{N}_1+\dot{N}_2$$ 

or 

$$\dot{N}_1=-\dot{N}_2$$

<div class="alert alert-block alert-info">
<b>Note:</b> the sine convention for mass entering and leaving the system is preserved by the conservation of mass.
</div>

#### (2) Energy Balance Expression
$$\dfrac{dU}{dt}=\sum_{k}\big(\dot{N}\underline{H}\big)_k+\dot{Q}+ \dot{W}$$

- steady state imples $\dfrac{d}{dt}=0$, so $\dfrac{dU}{dt}=0$
- there is a single input stream (1) and a single output stream (2) such that $\sum_k\big(\dot{N}\underline{H}\big)_k=\dot{N}_1\underline{H}_1+\dot{N}_2\underline{H}_2$
- the expansion is adiabatic such that $\dot{Q}=0$
- there is no boundary work such that $\dot{W_b}=0$

The **Conservation of Energy** thus simplifies to:

$$0=\dot{N}_1\underline{H}_1+\dot{N}_2\underline{H}_2 + W_s$$


#### (1+2) Combined Expressions

We can now combine the above expressions by setting $\dot{N}_1=-\dot{N}_2$ from (1) and subsituting into (2)

$$W_s=-\big(\dot{N}_1\underline{H}_1-\dot{N}_1\underline{H}_2\big)$$

or 

$$W_s=\dot{N}_1\big(\underline{H}_2-\underline{H}_1\big)$$

We conclude from this exercise that the steady state, adiabatic expansion across a turbine causes a change in enthalpy of the flowing fluid. 

- $\Delta\underline{H}>0$ when work is done on the system (this is the work of compression)
- $\Delta\underline{H}<0$ when work is done by the system (this is the result of expansion when work is done by the system)

### Transient, adiabatic expansion of a gas inside a piston

This derivation proceeds as above, except now there is **boundary work** and system is **closed**

#### (1) Mass Balance Expression

$$\dfrac{dN}{dt}=\sum_{k} \big(\dot{N}\big)_k$$

- there is no mass flow across the system boundary such that $\dot{N}_k=0$

The **Conservation of Mass** thus simplifies to:

$$\dfrac{dN}{dt}=0$$ 

and we correctly conclude that the mass change within the piston is constant (integrating the ordinary differential equation with respect to time confirms this conclusion). Integration assuming an initial condition $N(t=0)=N_0$ yields:

$$N=N_0$$

#### (2) Energy Balance Expression
$$\dfrac{dU}{dt}=\sum_{k}\big(\dot{N}\underline{H}\big)_k+\dot{Q}+ \dot{W}$$

- there is no mass flow across the system boundary such that $\big(\dot{N\underline{H}}\big)_k=0$
- the expansion is adiabatic such that $\dot{Q}=0$
- there is no shaft such that $\dot{W_s}=0$


The **Conservation of Energy** thus simplifies to:

$$\dfrac{dU}{dt}=-P^{ext}\dfrac{dV}{dt}$$

If we consider the result above,

- expansion where $\Delta V>0$, yields $\Delta U <0$
- compresssion where $\Delta V>0$, yields $\Delta U > 0$
