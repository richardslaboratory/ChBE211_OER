# Gaseous Mixtures

Gaseous mixtures are generally considered to be easier to quantitatively predict the state behavior as compared with liquid mixtures. This is because gases are generally more expanded than liquids. Therefore, their interactions are generally less frequent than in liquids and of less consequence to their material properties.

While there are many quantitative theories that seek to describe the state of a mixture at a fixed temperature and pressure as a function of composition, we will focus on two methods:

1. Ideal Gas Mixtures

2. Cubic Mixture Equations of State

## Ideal Gas Mixtures

**Ideal gas mixtures** are mixtures which obey the *ideal gas law* which for the purposes of this discussion, we write as:

$$PV=NRT$$

An ideal gas mixture is a mixture where the gas molecules are so expanded that they neither interact with other species in the mixture nor themselves. We can compute the properties of an ideal gas mixture by assuming that each species in the mixture contributes to to the total pressure proportionally to the number of moles of that speices or:

$$PV=RT\sum_i^C N_i$$

In addition to obeying the ideal gas law, ideal gas mixtures have a number of important properties. 

1. $\bar{V}_i=\underline{V}_i$

2. $\bar{U}_i=\underline{U}_i$

3. $\bar{H}_i=\underline{H}_i$

Another important property of ideal gas mixtures is their **Partial Pressure**, $P_i$ is given as the product of the total pressure and the mole fraction (this can be derived from eq (ref needed) above). Mathematically, we say that  

$$P_i=y_iP=\dfrac{N_iRT}{V}$$

<div class="alert alert-block alert-info">
<b>Note:</b> To denote mole fraction of a gaseous species, it is tradition to use y instead of x.
</div>

Because there is no difference in internal energy or density of an ideal gas containing $N$ moles of an ideal gas versus $N=\sum_i N_i$ moles of mixture species, it is tempting to think that the formation of an ideal gas mixture has no energetic conequence. However, where ideal gas mixtures become distinguishable from ideal gases is their *Entropy*. 

The entropy change upon forming an ideal gas mixture is given as:

$$\Delta_{mix}\underline{S}=-R\sum_i^{C}y_i\ln{y_i}$$

This equation, sometimes called the Stirling Approximation resembles the [multiplicity equation](https://en.wikipedia.org/wiki/Multiplicity_(statistical_mechanics) given by Boltzmann. This is not an accident. What this equation states is that ideal gas mixtures are mixtures where the species are thoroughly mixed such that all atoms in the mixture have an equal likelihood to be present within the volume.

Based on the above, the thermal properties of Ideal Gas Mixtures can be summarized:

* $\underline{U}^{IGM}(T,\underline{x})=\sum_i y_i\underline{U}_i^{IG}(T)$


* $\underline{V}^{IGM}(T,P,\underline{x})=\sum_i y_i\underline{V}_i^{IG}(T,P)$


* $\underline{H}^{IGM}(T,P,\underline{x})=\sum_i y_i\underline{H}_i^{IG}(T,P)$


* $P(T,\underline{V})=\sum_i P_i(T,V)$


* $\underline{S}^{IGM}(T,P,\underline{x})=\sum_i y_i\underline{S}_i^{IG}(T,P)-R\sum_i y_i \ln{y_i}$

## Non-ideal Gas Mixtures

At higher pressures and densities, significant deviation from Ideal Gas Mixture behavior is frequently encountered. Just like with single component fluids, these scenarios are treated using virial expansions of the equation of state, but modified for mixtures. 

### Virial Equation of State for Mixtures

At low pressures, the 2nd-order virial expansion proves to be a good approximation:

$$\dfrac{P\underline{V}}{RT}=1+\dfrac{B_{mix}(T,\underline{y})}{\underline{V}}$$

This equation is identical in form to the single component expression except now $B_{mix}$ depends on the composition of the mixture. Various **mixing rules** are used to describe the composition dependence of $B_{mix}$ but from statistical mechanics, we have:

$$B_{mix}(T,\underline{y})=\sum_i\sum_j y_i y_j B_{ij}(T)$$

where $B_{ij}=B_{ji}$

For a **Binary Mixture** this simplifies to:

$$B_{mix}(T,\underline{y})=y_1^2B_1(T)+2y_1 y_2 B_{12}(T)+y_2^2B_2(T)$$

We can see that by using this simple mixing rule, a new parameter has appeared, $B_{12}(T)$, which is a property unique to a given mixture.

### Cubic Equations of State for Mixtures

At higher pressures, the cubic equations of state must be used. Similar to the Virial Expansion, the cubic equations of state retain their same functional form with new mixing rules that describe the attractive and repulsive interaction parameters $a$ and $b$. As an example, we can write the mixture *Peng-Robinson* Equation of state as:

$$P=\dfrac{RT}{\underline{V}-b_{mix}}-\dfrac{a_{mix}(T)}{\underline{V}(\underline{V}+b_{mix})+b_{mix}(\underline{V}-b_{mix})}$$

using the following mixing rules:

1. $a_{mix}=\sum_i\sum_j y_i y_j a_{ij}$

where $a_{ij}=\sqrt{a_{ii} a_{jj}}(1-k_{ij})=a_{ji}$ and $k_{ij}$ is a dimensionless quantity defined as the *binary interaction parameter* that is unique for each pair in the mixture.

2. $b_{mix}=\sum_i y_i b_i$

## Thermodynamic Properties of Gaseous Mixtures

With the mixing rules defined above, it is possible to estimate thermodynamic properties of Gaseous mixtures with good accuracy using the same approach defined in Ch 4. We simply need to:

1. Define the composition

2. Define the state

3. Calculate the density


Using that density, we can then carry through the integrals in Ch 4. and evaluate property changes upon mixing for gas mixtures of various types.


## Lets see a numerical example of how these mixing rules work

Let's take the case of ethanol and water. 

The critical properties of ethanol are:

- $T_c = 516.25$ $K$

- $P_c = 6.390$ $MPa$

- $\rho_c = 280$ $kg/m^3$

- $MW = 46.1$ $g/mol$

- $w = 0.644$

The critical properties of water are:

- $T_c = 647.3$ $K$

- $P_c = 22.048$ $MPa$

- $\underline{V}_c = 0.056$ $m^3/kmol$

- $MW = 18.015$ $g/mol$

- $w = 0.344$

The interaction parameter $\kappa_{12} = -0.1$

### In the example below

In the example below, we plot the isothermal data using the PR_EOS for the mixture of ethanol and water. We see that with increasing mole fraction of water in the mixture, the mixture becomes less volatile resulting in the non-monotonic change in pressure as a function of molar volume at higher water fractions. 

This example highlights a few things:

1. The additional degree of freedom associated with mixtures. For any isotherm in the one phase region, selection of a Pressure results in a unique molar volume only if the composition is specified

2. In the fully expanded limit, $\underline{V}\rightarrow\infty$, the mixture obeys the ideal gas law

3. The 2-phase region of the VLE, we find  

```python
import numpy as np
import matplotlib.pyplot as plt

water = {'Tc':647.3, 'Pc':22.048*10**6, 'Vc':0.056/1000, 'w':0.344}
ethanol = {'Tc':516.25, 'Pc':6.390*10**6, 'Vc':46.1/280/1000, 'w':0.644}

k12 = -0.1
R = 8.314
##composition dependence of the PR_EOS
def Pr_EOS(T,x1):
    
    ##calculate the PR constants for each species
    #species 1 is water
    #species 2 is ethanol
    
    kappa1 = 0.37464+1.54226*water['w']-0.26992*water['w']**2
    kappa2 = 0.37464+1.54226*ethanol['w']-0.26992*ethanol['w']**2
    alpha1 = 1 + kappa1*(1 - np.sqrt(T/water['Tc']))
    alpha2 = 1 + kappa2*(1 - np.sqrt(T/ethanol['Tc']))
    a1 = 0.45724*(R*water['Tc'])**2/water['Pc']*alpha1
    a2 = 0.45724*(R*ethanol['Tc'])**2/ethanol['Pc']*alpha2
    b1 = 0.07780*(R*water['Tc'])/water['Pc']
    b2 = 0.07780*(R*ethanol['Tc'])/ethanol['Pc']
    
    
    ##apply mixing rules
    b = x1*b1+(1-x1)*b2
    a = x1**2*a1 + (1-x1)**2*a2 + 2*x1*(1-x1)*np.sqrt(a1*a2)*(1-k12)
    V = np.logspace(np.log10(1.1*b),np.log10(1000*b),201)
    return V,R*T/(V-b) - a/(V*(V+b) + b*(V-b))


V = np.logspace(-5,1)
T = 310 #C
plt.figure(dpi=300,figsize=(3,3))
for x1 in np.arange(0,1.2,0.2):
    V,P = Pr_EOS(T+273.15,x1)
    plt.plot(V*10**3,P/10**5, label = '%3.2f'%x1)
    
plt.yscale('log')
plt.xscale('log')
plt.ylabel('P (bar)')
plt.xlabel('V (l/mol)')
plt.annotate(f'T = {T} C',xy=(0.05,0.05), xycoords = 'axes fraction')
plt.legend(frameon=False,fontsize=4)
```

## Recalling our definitions from Ch 6. 

$\quad d\underline{U}=C_v dT + \bigg[T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}} - P\bigg]d\underline{V}$

$\quad d\underline{S}=\dfrac{C_v}{T} dT + \bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}d\underline{V}$

$\quad d\underline{H}=C_p dT + \bigg[\underline{V} - T\bigg(\dfrac{\partial \underline{V}}{\partial T}\bigg)_{P}\bigg]dP$

$\quad d\underline{S}=\dfrac{C_p}{T} dT - \bigg(\dfrac{\partial \underline{V}}{\partial T}\bigg)_{P}dP$

You may be able to see that thermal variables can still be computed from the volumetric form of the E.O.S by taking the three step pathway through the ideal gas state described in Ch 6. However, now the thermal data will be sensitive to the mixture composition through the mixing variables $a_{mix}(x_1, T)$ and $b_{mix}(x_1)$.

Note that now in the ideal gas state step, I will need to utilize the properties of an ideal gas mixture for this computation, 

$$\bigg(\dfrac{\partial \underline{U}}{\partial T}\bigg)_{\underline{V}}^{IG} = \dfrac{\partial}{\partial T}\bigg|_{\underline{V}}\underline{U}
= \dfrac{\partial}{\partial T}\bigg|_{\underline{V}} \bigg[\sum_i y_i\underline{U}_i^{IG}(T)\bigg]$$

Integrating with respect to $T$ for a binary mixture in the ideal gas state from $T_1$ to $T_2$, we find:


$ \underline{U} = y_1 C_V^{IG,1}(T_2-T_1) + y_2 C_V^{IG,2}(T_2-T_1)$

$ \underline{H} = y_1 C_P^{IG,1}(T_2-T_1) + y_2 C_P^{IG,2}(T_2-T_1)$

$ \underline{S} = y_1 C_P^{IG,1}ln(T_2/T_1) + y_2 C_P^{IG,2}ln(T_2/T_1) - R[y_1 ln(y_1) + y_2 ln(y_2)]$

Note the additional entropy present due to the **excess entropy** of the ideal gas


and $\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}} = \dfrac{R}{\underline{V}-b_{mix}}-\dfrac{da_{mix}/dT}{\underline{V}(\underline{V}+b_{mix})+b_{mix}(\underline{V}-b_{mix})}$

and therefore:
$T\bigg(\dfrac{\partial P}{\partial T}\bigg)_{\underline{V}}-P = \dfrac{a_{mix} - Tda_{mix}/dT}{\underline{V}(\underline{V}+b_{mix})+b_{mix}(\underline{V}-b_{mix})}$
